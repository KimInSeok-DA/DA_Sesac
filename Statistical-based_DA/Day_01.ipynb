어떤 도로에 차량 신호등이 n개 있습니다.

모든 신호등은 항상 초록불 → 노란불 → 빨간불 순서로 반복되며, 각 신호의 지속 시간은 신호등마다 다릅니다.

시간은 1초부터 시작하며, 각 신호등은 처음에는 초록불 상태로 시작합니다.

이 도로에서는 가끔 정전이 일어나는데,

모든 신호등이 모두 노란불이 되면 정전이 발생한다는 사실이 밝혀졌습니다.

예를 들어 신호등이 2개이고, 각 신호등의 주기가 다음과 같다고 가정해 보겠습니다.

# 논리 순서

1. 입력에 따라 신호등 색깔과 순서 정의
2. G->Y->R 순서로 나열
3. 각각의 숫자만큼 패턴 반복
4. n 개의 신호등 모든 것이 Y 로 나타나는 구간 도출(x초)

```python
# 풀이
signals1 = [[2, 1, 2], [5, 1, 1]]                       # 13
signals2 = [[2, 3, 2], [3, 1, 3], [2, 1, 1]]            # 11
signals3 = [[3, 3, 3], [5, 4, 2], [2, 1, 2]]            # 193
signals4 = [[1, 1, 4], [2, 1, 3], [3, 1, 2], [4, 1, 1]] # -1

def solution(signals):
  time = 1
  MAX_TIME = 1500000                          ## 안겹칠 경우 : 모든 채점 통과를 위한 수치
  while time <= MAX_TIME :                    ## 무한 반복 방지
    all_yellow = True                         # '모든 신호등이 노란불이다' 를 기준으로 함
    for idx, signal in enumerate(signals):
      green = signal[0]
      yellow = signal[1]
      red = signal[2]
      total_cycle = sum(signal)
      remain = time % total_cycle              # 나누기를 통한 나머지를 활용해 현재 위치 확인 // [1/5] 로 나머지가 1이니 첫번째 위치(초록불 상태) 확인
      if remain == 0:                          # 나머지가 0인 경우는 맨 마지막 위치인 것
        remain = total_cycle                     #  [5/5] 는 나머지가 0이지만 마지막 위치(빨간불 상태) 확인
      if green < remain <= (green + yellow):   # 노란불 구간 검사: 초록불 초과 ~ (초록불 + 노란불) 이하인지 확인 //// 3초일때, 6초일때 노란불
        pass                                     # 조건에 맞으면 노란불이므로 아무것도 안 하고 넘어감 (다음 검사 진행)
      else:
        all_yellow = False                     # 신호등이 노란불이 아닌 경우에는 더이상 진행할 필요없음
        break                                    # for 문 탈출하고, time에 +1 함
    if all_yellow :
      return time
    time += 1
  return -1

print(solution(signals1))
print(solution(signals2))
print(solution(signals3))
print(solution(signals4))
```

# 층화 추출방식

대한민국 고등학생의 학업 성취도를 조사하기 위해 모집단을 [1학년 주머니], [2학년 주머니], [3학년 주머니]로 나눕니다.

그리고 1학년 주머니에서 100명, 2학년에서 100명, 3학년에서 100명을 각각 무작위로 뽑아 총 300명을 조사합니다.


층 내부(동질): 1학년 주머니 안에는 1학년들만 있으므로 내부 구성원의 성격이 비슷합니다.

층 간(이질): 1학년 주머니와 3학년 주머니는 서로 확연히 다른 성격을 가집니다.

# 집락 추출방식

대한민국 고등학생을 조사하려는데 전국을 다 돌아다니기 힘드니, 전국 고등학교 중 [A고등학교], [B고등학교], [C고등학교] 세 군데(덩어리)만 무작위로 뽑습니다.

그리고 선택된 A, B, C 고등학교의 전교생을 전부 조사해 버립니다. 나머지 학교는 조사하지 않습니다.


집락 내부(이질): A고등학교 안에는 1, 2, 3학년, 공부 잘하는 학생, 못하는 학생 등 다양한 사람들이 섞여 있어 모집단의 축소판처럼 복잡합니다.

집락 간(동질): A고등학교나 B고등학교나 대다수 평범한 일반계 고등학교라면 학교끼리의 전반적인 성격은 비슷비슷합니다.
