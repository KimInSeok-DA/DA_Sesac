# Matplotlib 실습

```python
# 실습 1-1 서브플롯 객체에서 플롯하기

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(4,3))   # figure : 스케치북 / figsize : 사이즈 조절

ax1 = fig.add_subplot(2,2,1)      # add_subplot : 스케치북을 쪼개서 1번 위치에 정리
ax2 = fig.add_subplot(222)        # 쉼표 생략 가능 / 2*2 사이즈로 (1번, 2번, 3번)위치에 정리
ax3 = fig.add_subplot(223)


# plt.plot : 그래프 그리기(지정되지 않으면 가장 최근에 만든 것에 그림)

plt.plot(np.random.randn(50).cumsum(), 'k--')                    # 50개의 숫자를 랜덤으로 뽑아서 누적해서 더하기 / 검은색(k) 점선(--)
ax1.hist(np.random.randn(100), bins=20, color='g', alpha=0.3)    # ax1 은 히스토그램(hist) / 20개의 구간으로 쪼개서, 초록(g), 투명도(30%)
ax2.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))  # ax2 는 산점도(scatter) / x값 30, y값은 30에다가 정규분포된 무작위 30을 곱한 것
plt.show() # 그래프 그리기
```

# size의 단위 설명

Matplotlib에서 figsize=(15, 5)의 크기 단위는 인치(Inch)

fig = plt.figure(figsize=(4,3))

4: 가로 4인치

3: 세로 3인치

그럼 모니터 화면(픽셀)으로는 얼마나 크게 나오는 걸까요?

여기서 실제 화면에 그려지는 픽셀 크기를 결정하는 바로 DPI(Dots Per Inch, 1인치당 들어가는 픽셀 수)이기 때문

Matplotlib의 기본 DPI 기본값은 100

따라서  코드는 컴퓨터 내부적으로 다음과 같이 계산되어 화면에 뿌려집니다.

가로 픽셀 = 4인치 $\times$ 100 DPI = 400 픽셀

세로 픽셀 = 3인치 $\times$ 100 DPI = 300 픽셀

400 $\times$ 300 사이즈로 그려짐

```python
# 실습 1-2 눈금, 라벨, 범례(p.87)

fig = plt.figure(figsize=(5,2))         # 사이즈 정하기
ax = fig.add_subplot(1,1,1)             # 사이즈를 쪼개지 않고, 하나로 만들기
ax.plot(np.random.randn(1000).cumsum()) # 0~1000개를 연속으로 더하기(우상향 그래프)

# x축 그래프의 눈금 설정
ticks = ax.set_xticks([0,250,500,750,1000]) # x축 눈금 고정
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small') # 그걸 문자열로 바꾸고 각도, 글씨 크기 조절

# 제목 설정

ax.set_title('My first matplotlib plot') # 메인 제목 설정
ax.set_xlabel('Stages')                  # x축 아래 설명 설정

fig.subplots_adjust(bottom=0.3)          # 여백 30% 적용
plt.show()

plt.close('all')                         # 스케치북창 닫기
```

```python
# 실습 3 plotly 패키지 실습(예제)

import pandas as pd
import plotly.graph_objects as go                                      # 동적 그래프 그리기

df = pd.DataFrame({
    'Source': ['창고','창고','서울허브','서울허브','부산허브'],
    'Target': ['서울허브','부산허브','서울배송','경기배송','부산배송'],
    'Value': [1000,500,600,400,500]
})

# 노드 생성
labels = list(
    pd.concat([df['Source'], df['Target']]).unique()
)

# 노드 인덱스 매핑
mapping = {label:i for i,label in enumerate(labels)}

# Sankey용 데이터
source = df['Source'].map(mapping)
target = df['Target'].map(mapping)
value = df['Value']

fig = go.Figure(go.Sankey(
    node=dict(
        label=labels,
        pad=15,
        thickness=20
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
))

fig.update_layout(
    title="물류 데이터 Sankey Diagram",
    font_size=12
)

fig.show()
```

# fl2000.txt Data EDA

1. 각 후보의 총득표수를 비교할 수 있는 그래프를 그리시오
2. 각 후보들의 top10 의 county 를 비교할 수 있는 그래프를 그리시오
3. 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
(단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)
4. Bush 가 Gore 보다 많은 득표를 한 county 별로 Bush, Gore 의 득표수를 비교할 수 있는 그래프를 그리시오
5. Bush, Gore 득표수가 많은 county 중에 공통으로 존재하는 county 10곳에 대한 득표수를 비교할 수 있는 그래프를 그리시오

```python
# 실습 1 각 후보의 총 득표수를 비교할 수 있는 그래프 그리기

import seaborn as sns

# 후보자 컬럼만 추출 (5번째 컬럼인 'Bush'부터 끝까지)
candidates = fl2000.columns[4:]

# 각 후보자의 총 득표수 합산 하고 내림차순 정렬
total_votes = fl2000[candidates].sum().sort_values(ascending=False)  # 정렬

# seaborn 스타일
sns.set_theme(style="whitegrid")

# 사이즈 선
plt.figure(figsize=(12, 6))

# 바 차트(Bar Chart) 그리기
sns.barplot(x=total_votes.index, y=total_votes.values,hue=total_votes.index, palette='Spectral', legend=False)

# 그래프 디테일 설정 (제목, 축 이름, 눈금 설정 등)
plt.title('Total Votes', fontsize=16, fontweight='bold')
plt.xlabel('Candidates', fontsize=12)
plt.ylabel('Total Votes', fontsize=12)
plt.xticks(rotation=45)                       # x축 이름 회전
plt.ticklabel_format(style='plain', axis='y')


'''
2000년 대선에서 플로리다 주 표차이는 1784표
'''

plt.show()
```

```python
# 실습 1 각 후보의 총 득표수를 비교할 수 있는 그래프 그리기
# + Bush와 Gore 막대에만 숫자 달기

candidates = fl2000.columns[4:]
total_votes = fl2000[candidates].sum().sort_values(ascending=False)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=total_votes.index, y=total_votes.values, hue=total_votes.index, palette='Spectral',legend=False) # 경고문 발생 주범

# enumerate를 쓰면 i에는 순번(0, 1, 2...), candidate에는 후보 이름이 들어갑니다.
for i, candidate in enumerate(total_votes.index):

    #  Bush 랑 Gore 만 숫자 입력
    if candidate in ['Bush', 'Gore']:

        # ax.text()를 이용해 원하는 x, y 좌표에 직접 글씨를 씁니다.
        ax.text(
            x=i,                                 # 가로 위치 (i번째 막대의 중앙)
            y=total_votes.values[i],             # 세로 위치 (막대의 높이 = 득표수)
            s=f'{total_votes.values[i]:,.0f}',   # 적을 텍스트 (천 단위 콤마 추가)
            ha='center',                         # 가로 중앙 정렬
            va='bottom',                         # 세로 정렬: 막대 선의 바로 위(bottom)에 안착
            fontsize=12,
            fontweight='bold',
            color='black'
        )

# 4. 그래프 디테일 설정
plt.title('Florida 2000 Presidential Election - Total Votes', fontsize=16, fontweight='bold')
plt.xlabel('Candidates', fontsize=12)
plt.ylabel('Total Votes', fontsize=12)
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')

# 출력
plt.tight_layout()
plt.show()
```

```python
# 실습 3-2 각 후보들의 top10 의 county 를 비교할 수 있는 그래프를 그리시오

# 12명 후보 전체 이름 가져오기 (전체 득표수 기준 내림차순 정렬)
candidates = fl2000.columns[4:]
all_candidates = fl2000[candidates].sum().sort_values(ascending=False).index

# 도화지 세팅: 4행 3열 (4x3 = 12개) 사이즈의 거대한 액자 만들기
sns.set_theme(style="whitegrid")

# 도화지 크기(figsize)를 12개
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 12))
axes = axes.flatten()  # 4x3 행렬 형태의 액자를 1차원 리스트로 피기.

# 반복문(for)을 돌면서 4명의 후보 각각의 Top 10 카운티 그래프 그리기
for i, candidate in enumerate(all_candidates):
    # 해당 후보의 득표수를 기준으로 내림차순 정렬 후 상위 10개만 자르기
    top10_counties = fl2000[candidate].sort_values(ascending=False).head(10)

    # seaborn 가로형 막대 그래프 그리기
    # 카운티 이름이 길기 때문에 y축에 카운티(index)를 넣고, x축에 득표수(values)를 넣기
    sns.barplot(
        x=top10_counties.values,
        y=top10_counties.index,
        hue=top10_counties.index,
        palette='Spectral',
        legend=False,
        ax=axes[i]                 # i번째 액자에 그리기
    )

    # 각 액자별 디테일 설정
    axes[i].set_title(f'Top 10 Counties for {candidate}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Total Votes', fontsize=11)
    axes[i].set_ylabel('County', fontsize=11)
    axes[i].ticklabel_format(style='plain', axis='x')  # 득표수 지수표현(1e6) 방지


plt.tight_layout()
plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 1. 산점도 : 각 카운티별로 두 후보가 받은 득표수의 분포와 상관관계를 점으로 찍기

fl2000.plot(kind='scatter', x='Bush', y='Gore', color='purple', alpha=0.6, figsize=(8,6))
plt.title('Gore VS Bush Votes Scatter')

'''
y=x 를 기준으로
점 하나하나는 플로리다주의 67개 카운티(지역)
선거의 승패를 결정지은 '메가 카운티(최대 격전지)'
실제 데이터를 뜯어보면 이 점들은 인구가 밀집된 Miami-Dade, Broward, Palm Beach, Pinellas 카운티입니다.

특정 지역(그래프 우측 상단의 점들)에서 두 후보의 격차가 얼마나 크게 벌어졌는지,
즉 '승부처'가 된 초대형 카운티들의 이상치(Outlier)를 한눈에 볼 수 있습니다.
'''

plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 2. 득표차이 추가 컬럼 생성

# Bush 득표수에서 Gore 득표수를 뺀 '격차(Margin)' 컬럼 생성
fl2000['Margin'] = fl2000['Bush'] - fl2000['Gore']

# 격차 데이터를 20개의 막대(bins)로 쪼개어 히스토그램(hist) 생성
fl2000['Margin'].plot(kind='hist', bins=20, color='teal', edgecolor='black', figsize=(8,6))
plt.title('Margin [Bush - Gore] Distribution')
plt.axvline(0, color='red', linestyle='dashed', linewidth=2)                                  # 0을 기준으로 기준선 추가

'''

각 지역에서 Bush가 Gore를 몇 표 차이로 이겼는지(또는 졌는지) '표 차이' 컬럼을 만들고, 그 분포 확인.

중심(0표 차이) 부근에 얼마나 많은 카운티가 몰려 있는지 보여줍니다.

2000년 플로리다 대선이 역사상 가장 치열했던 초박빙 선거였음을 직관적으로 증명

'''

plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 3. 무효표(Under)와 중복표(Over)의 이상치 탐지

fl2000[['under', 'over']].plot(kind='box', color='blue', figsize=(8,6))
plt.title('Undervotes and Overvotes Distribution')
plt.ylabel('Number of Votes')

'''

기계가 읽지 못한 표(under)와 두 명을 찍은 표(over)의 분포

당시 펀치 카드 투표기 문제로 특정 지역에서 기형적으로 무효표가 쏟아졌던 사태를 파악할 수 있습니다.

상자(Box) 위로 멀리 떨어진 점(Outlier)들이 바로 논란이 되었던 문제의 카운티

'''

plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 4. 투표 기계(Technology)별 사용 비율

fl2000['Total_Votes'] = fl2000[fl2000.columns[4:]].sum(axis=1)

# 기계 방식(technology)별로 그룹화하여 파이 차트(pie) 생성
tech_votes = fl2000.groupby('technology')['Total_Votes'].sum()
tech_votes.plot(kind='pie', autopct='%1.1f%%', cmap='Set2', figsize=(8,6))
plt.title('Voting Technology')
plt.ylabel('') # 파이차트는 y라벨이 지저분해져서 지워줍니다.

'''

플로리다 전체 표심 중 어떤 투표 기계 방식이 가장 많이 쓰였는지 파이를 쪼개어 확인합니다.

당시 재검표 논란의 핵심이었던 펀치 카드(Votomatic 등) 방식과 광학 스캐너(Optical) 방식의 점유율을 통해,

시스템 오류가 선거 전체에 미칠 수 있었던 영향력을 가늠할 수 있습니다.

'''
# 3번과 4번을 섞으면 보기 더 편하지 않을까?

plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 5. 투표 기계(Technology)별 무효표(Under)와 중복표(Over)의 이상치 탐지

plot_data = fl2000[['technology', 'under', 'over']].reset_index()
melted_data = plot_data.melt(id_vars=['county', 'technology'],
                             value_vars=['under', 'over'],
                             var_name='Vote_Type',
                             value_name='Count')

# 스타일 세팅
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 8))

# 박스 플롯(Box Plot) 그리기
# x축엔 '투표 방식(technology)', y축엔 '표 개수(Count)', 색상(hue)은 '무효/중복(Vote_Type)'으로 나눕니다.
sns.boxplot(data=melted_data, x='technology', y='Count', hue='Vote_Type', palette='Set2')

# 그래프 디테일 설정
plt.title('Florida 2000: Under/Over Votes by Technology (The Smoking Gun)', fontsize=18, fontweight='bold')
plt.xlabel('Voting Technology', fontsize=14)
plt.ylabel('Number of Invalid Votes (Under/Over)', fontsize=14)
plt.xticks(fontsize=12)
plt.ticklabel_format(style='plain', axis='y')

# 출력
plt.tight_layout()

'''

펀치 카드(Votomatic 등) 방식에서 중복표가 많이 발생한 것으로 보임

'''
plt.show()
```

```python
# 실습 3-3 판다스의 데이터 프레임에서 제공하는 그래프를 이용하여 유용한 그래프를 그리시오
# (단, 5개이상, 그래프가 나타내는 인사이트가 있어야함)

# 6. Top 10 지역의 핵심 후보 점유율

# 총 득표수 기준 상위 10개 카운티 추출
top10 = fl2000.sort_values(by='Total_Votes', ascending=False).head(10)

# 주요 후보 3명만 뽑아서 누적(stacked=True) 막대그래프(bar) 생성
top10[['Bush', 'Gore', 'Nader']].plot(kind='bar', stacked=True, color=['#d62728', '#1f77b4', '#2ca02c'], figsize=(12,6))
plt.title('5. Top 10 Counties: Vote Share (Bush vs Gore vs Nader)')
plt.xticks(rotation=45, ha='right')

'''

가장 표가 많이 나온 Top 10 카운티에서,
주요 후보 3명(Bush, Gore, Nader)의 득표수를 누적하여 보여줌.

단순히 막대가 나란히 있는 것보다, 그중 파란색(Gore)과 빨간색(Bush)이 파이를 어떻게 나눠 졌는지를 가장 직관적으로 보여줍니다.

'''

plt.show()
```

```python
# 실습 3-4 Bush 가 Gore 보다 많은 득표를 한 county 별로 Bush, Gore 의 득표수를 비교할 수 있는 그래프를 그리시오

# Bush가 Gore보다 득표수가 많은 카운티만 필터링
# fl2000['Bush'] > fl2000['Gore'] 부분은 True/False를 반환하고, 이걸로 자르기

bush_wins = fl2000[fl2000['Bush'] > fl2000['Gore']]

# 필요한 두 후보의 데이터만 뽑고, 보기 좋게 정렬 (Bush 득표수 기준 내림차순)
compare_data = bush_wins[['Bush', 'Gore']].sort_values(by='Bush', ascending=False)

# 그룹화된 막대 그래프(Grouped Bar Chart)

# color 옵션으로 Bush은 빨간색, Gore은 파란색 계열로 직관성 향상
ax = compare_data.plot(kind='bar', figsize=(20, 8), color=['#d62728', '#1f77b4'], width=0.8)

# 그래프 디테일 설정
plt.title('Bush won against Gore', fontsize=18, fontweight='bold')
plt.xlabel('County', fontsize=14)
plt.ylabel('Votes', fontsize=14)
plt.xticks(rotation=45, ha='right')           # x축 지역 이름이 겹치지 않게 기울이고 우측 정렬
plt.ticklabel_format(style='plain', axis='y') # y축 수치 지수표기법(1e6) 방지
plt.legend(title='Candidate', fontsize=12)    # 범례 표시
plt.grid(axis='y', linestyle='--', alpha=0.7) # 가독성을 위한 가로선 추가

# 여백 자동 조정 및 출력
plt.tight_layout()
plt.show()
```

```python
# 실습 3-5 Bush, Gore 득표수가 많은 county 중에 공통으로 존재하는 county 10곳에 대한 득표수를 비교할 수 있는 그래프를 그리시오

# Bush와 Gore 각각의 득표수 상위 15개 지역(County) 추출 후 집합(Set)으로 만들기
# top 10개씩만 뽑으면 공통 지역이 10개가 안 될 수 있어서 넉넉히 뽑기

top_bush = set(fl2000['Bush'].sort_values(ascending=False).head(30).index)
top_gore = set(fl2000['Gore'].sort_values(ascending=False).head(30).index)

# 교집합(&) 연산자로 두 후보 모두에게 득표수가 많은 공통 지역 추출
common_counties = list(top_bush & top_gore)

# 공통 지역들 중에서, 두 후보의 합산 득표수가 가장 많은 '진짜 핵심 지역 Top 10' 추려내기
fl2000['Bush_Gore_Total'] = fl2000['Bush'] + fl2000['Gore'] # 임시로 합산 파이(Total) 컬럼 만들기
top10_common = fl2000.loc[common_counties].sort_values(by='Bush_Gore_Total', ascending=False).head(10)

# 시각화를 위해 Bush와 Gore 컬럼만 잘라오기
plot_data = top10_common[['Bush', 'Gore']]

# 그룹화된 막대 그래프(Grouped Bar Chart) 그리기
ax = plot_data.plot(kind='bar', figsize=(14, 7), color=['#d62728', '#1f77b4'], width=0.7)

# 그래프 디테일 설정
plt.title('Top 10 Common Counties for Bush & Gore', fontsize=16, fontweight='bold')
plt.xlabel('County', fontsize=14)
plt.ylabel('Total Votes', fontsize=14)
plt.xticks(rotation=45, ha='right')           # 글씨 안 겹치게 45도 기울이기
plt.ticklabel_format(style='plain', axis='y') # y축 수치 지수표현 방지
plt.legend(title='Candidate', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 출력
plt.tight_layout()
plt.show()
```

```python
# 실습 4 tips 파일로 실습

os.getcwd()
tips = pd.read_csv('tips.csv')
tips
```

```python
# 실습 4-1 새로운 속성 만들기

tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips
```

```python
# 실습 4-2 기준점을 토대로 상위 도출하는 함수 만들기

def top(df, n=5, column='tip_pct'):       # tip_pct 기준으로
    return df.sort_values(by=column)[-n:] # 디폴트값인 오름차순 정렬로 뒤에서부터 n개 출력

top(tips, n=6)
```

```python
# 실습 4-3 정렬하고 함수 적용

# smoker 의 값 기준으로 분류 / top 함수 적용, include_groups=False : 그룹핑한 기준 컬럼은 결과에서 제외해 줘
tips.groupby('smoker').apply(top, include_groups=False)
```

```python
# 실습 4-3 정렬하고 함수 적용

# smoker 와 day 의 값 기준으로 분류
# 정렬 기준은 top 함수, 각 1개씩 뽑고, total_bill로 기준
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill', include_groups=False)
```

```python
# 실습 5-1 보려는 값만 추리기

tips.pivot_table(index=['sex','smoker'], values=['size', 'tip','tip_pct', 'total_bill'])
```

```python
# 실습 5-2 기준 조정

tips.pivot_table(['tip_pct', 'size'], index=['sex', 'day'], columns='smoker')
```
