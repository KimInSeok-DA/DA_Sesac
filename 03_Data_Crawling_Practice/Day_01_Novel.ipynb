```python
# 웹소설 사이트 크롤링

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://novel.naver.com/webnovel/weekday"

user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36"
headers = {"User-Agent": user_agent}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

print("네이버 웹소설 페이지 가져오기 성공!")
```

```python
# 웹소설 목록 찾기

novel_list = soup.find_all("li", class_="item")

print("찾은 웹소설 개수:", len(novel_list))
```

```python
# 소설 제목, 링크, 관심수 뽑기

records = []

for novel in novel_list:
    try:
        # 1. 웹소설 제목 추출 
        title_tag = novel.find("p", class_="title") or novel.find("span", class_="title") or novel.find("strong")
        title = title_tag.get_text(strip=True) if title_tag else "제목 없음"
        
        # 2. 웹소설 링크 추출 
        a_tag = novel.find("a")
        if a_tag and a_tag.has_attr("href"):
            link = "https://novel.naver.com" + a_tag["href"]
        else:
            link = "링크 없음"
            
        # 3. 관심 수 추출
        count_tag = novel.find("span", class_="count") or novel.find("span", class_="score") or novel.find("em")
        count = count_tag.get_text(strip=True) if count_tag else "0"
        
        # 바구니에 저장
        records.append([title, link, count])
        
    except Exception as e:
        # 혹시 에러가 나면 패스하고 다음 걸로 진행
        continue

print("개수:", len(records))
records[:3]
```

```python
# CSV 파일 저장

df_novels = pd.DataFrame(records, columns=["제목", "링크", "관심 수"])

output_path = "네이버 웹소설 통합랭킹.csv"
# df_novels.to_csv(output_path, index=False, encoding="utf-8-sig")

print("저장 완료:", output_path)
df_novels.head(5) # 상위 10개 표 형태로 출력
```
