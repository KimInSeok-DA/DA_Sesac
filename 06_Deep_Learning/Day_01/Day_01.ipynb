## 1. нҷңм„ұнҷ” н•ЁмҲҳ (Activation): ReLU vs Sigmoid

нҷңм„ұнҷ” н•ЁмҲҳлҠ” мқҙм „ мёөм—җм„ң л„ҳм–ҙмҳЁ лҚ°мқҙн„°лҘј м–ҙл–Ө нҳ•нғңлЎң ліҖнҷҳн•ҙм„ң л„ҳкёём§Җ кІ°м •н•ҳлҠ” м—ӯн• мқ„ н•©лӢҲлӢӨ.

### 1) Sigmoid (мӢңк·ёлӘЁмқҙл“ң)

- **нҠ№м§•**: лӘЁл“  мҲ«мһҗлҘј 0кіј 1 мӮ¬мқҙмқҳ л¶Җл“ңлҹ¬мҡҙ кіЎм„ мңјлЎң м••м¶•. мқҢмҲҳлҠ” 0м—җ к°Җк№қкІҢ, м–‘мҲҳлҠ” 1м—җ к°Җк№қкІҢ л§Ңл“ӯлӢҲлӢӨ.
- **мҡ©лҸ„**: кІ°кіјк°’мқҙ 0~1 мӮ¬мқҙмқҳ 'нҷ•лҘ 'лЎң лӮҳмҷҖм•ј н•ҳлҠ” мқҙм§„ л¶„лҘҳ.
- **лӢЁм җ (мқҖлӢүмёөм—җм„ң м•Ҳ м“°лҠ” мқҙмң )**: мӢ кІҪл§қмқҙ к№Ҡм–ҙм§ҲмҲҳлЎқ к°’мқ„ кі„мҶҚ 0~1 мӮ¬мқҙлЎң м§“лҲ„лҘҙлӢӨ ліҙлӢҲ, лӮҳмӨ‘м—җлҠ” лҚ°мқҙн„°мқҳ нҠ№м§•(кё°мҡёкё°)мқҙ нқ¬лҜён•ҙм ём„ң н•ҷмҠөмқҙ л©Ҳм¶°лІ„лҰ¬лҠ” **кё°мҡёкё° мҶҢмӢӨ(Vanishing Gradient)** л°ңмғқ.

### 2) ReLU (л җлЈЁ, Rectified Linear Unit)

- **нҠ№м§•**: лӢЁмҲң. мқҢмҲҳл©ҙ к·ёлғҘ 0мңјлЎң м—Ҷм• лІ„лҰ¬кі , м–‘мҲҳл©ҙ к·ё мҲ«мһҗ к·ёлҢҖлЎң нҶөкіјмӢңнӮҙ (`max(0, x)`).
- **мҡ©лҸ„**: мқҖлӢүмёө(Hidden Layer)мқҳ н‘ңмӨҖ(Standard). нҠ№лі„н•ң мқҙмң к°Җ м—ҶлӢӨл©ҙ ReLUлҘј мӮ¬мҡ©.
- **мһҘм җ**: кі„мӮ° мҶҚлҸ„к°Җ м—„мІӯлӮҳкІҢ л№ лҘҙкі , кё°мҡёкё° мҶҢмӢӨ л¬ём ңлҘј мҷ„лІҪн•ҳкІҢ н•ҙкІ°н•ҳм—¬ мӢ кІҪл§қмқ„ мҲҳмӢӯ, мҲҳл°ұ мёөмңјлЎң мҢ“мқ„ мҲҳ мһҲкІҢ н•Ё.

## 2. мҶҗмӢӨ н•ЁмҲҳ (Loss): MSE vs Binary Crossentropy л“ұ

мҶҗмӢӨ н•ЁмҲҳлҠ” лӘЁлҚёмқҙ м •лӢөмқ„ л§һнһҲкі  лӮҳм„ң мұ„м җн•ҳлҠ” мҳӨм°Ё кі„мӮ°лІ•мһ…лӢҲлӢӨ.

### 1) MSE (Mean Squared Error, нҸүк·  м ңкіұ мҳӨм°Ё)

- **мӮ¬мҡ© мӢңкё°**: нҡҢк·Җ л¬ём ң (м§‘к°’ мҳҲмёЎ, мҳЁлҸ„ мҳҲмёЎ л“ұ кө¬мІҙм Ғмқё м—°мҶҚлҗң 'мҲ«мһҗ'лҘј л§һнһҗ л•Ң)
- **нҠ№м§•**: мӢӨм ң м •лӢө мҲҳм№ҳмҷҖ лӘЁлҚёмқҙ мҳҲмёЎн•ң мҲҳм№ҳмқҳ м°ЁмқҙлҘј м ңкіұн•ҳм—¬ нҸүк· . мҳӨм°Ёк°Җ нҒҙмҲҳлЎқ нҺҳл„җнӢ°лҘј кё°н•ҳкёүмҲҳм ҒмңјлЎң нҒ¬кІҢ мӨҚлӢҲлӢӨ.

### 2) Binary Crossentropy (мқҙм§„ көҗм°Ё м—”нҠёлЎңн”ј)

- **мӮ¬мҡ© мӢңкё°**: мқҙм§„ л¶„лҘҳ л¬ём ң (м•” м–‘м„ұ/м•…м„ұ, мҠӨнҢёл©”мқј O/X л“ұ 2к°ң мӨ‘ н•ҳлӮҳлҘј кі лҘј л•Ң)
- **нҠ№м§•**: нҷ•лҘ  кё°л°ҳ. м •лӢөмқҙ 1(м•”)мқёлҚ° лӘЁлҚёмқҙ 0.1(10%)мқҙлқјкі  мҳҲмёЎн–ҲлӢӨл©ҙ нҒ° нҺҳл„җнӢ°лҘј, 0.9(90%)лқјкі  мҳҲмёЎн–ҲлӢӨл©ҙ мһ‘мқҖ лІҢм җмқ„ мӨҚлӢҲлӢӨ.

### 3) Categorical Crossentropy (лӢӨмӨ‘ көҗм°Ё м—”нҠёлЎңн”ј) вҖ” лІҲмҷё

- **мӮ¬мҡ© мӢңкё°**: лӢӨмӨ‘ л¶„лҘҳ л¬ём ң (к°ң/кі м–‘мқҙ/мғҲ мӨ‘ н•ҳлӮҳ л§һнһҲкё°, м•„мқҙлҰ¬мҠӨ кҪғ 3мў…лҘҳ л¶„лҘҳ л“ұ м •лӢөмқҙ 3к°ң мқҙмғҒмқј л•Ң)

## 3. мөңм Ғнҷ” м•Ңкі лҰ¬мҰҳ (Optimizer): SGD vs Adam

мөңм Ғнҷ” м•Ңкі лҰ¬мҰҳмқҖ мҶҗмӢӨ н•ЁмҲҳлЎң кі„мӮ°лҗң мҳӨм°ЁлҘј л°”нғ•мңјлЎң м–ҙл–»кІҢ н•ҳл©ҙ к°ҖмһҘ л№ лҘҙкі  м•Ҳм „н•ҳкІҢ мҳӨм°ЁлҘј 0(мөңм Ғмқҳ к°ҖмӨ‘м№ҳ)мңјлЎң мӨ„м—¬лӮҳк°Ҳм§Җ кІ°м •н•ҳлҠ” м—”м§„мһ…лӢҲлӢӨ.

### 1) SGD (Stochastic Gradient Descent, нҷ•лҘ м Ғ кІҪмӮ¬ н•ҳк°•лІ•)

- **нҠ№м§•**: к°ҖмһҘ кё°ліём Ғмқҙкі  кі м „м Ғмқё л°©мӢқ. нҳ„мһ¬ м„ң мһҲлҠ” мң„м№ҳм—җм„ң кІҪмӮ¬(кё°мҡёкё°)к°Җ м ңмқј к°ҖнҢҢлҘё мӘҪмңјлЎң м§Җм •лҗң лӢЁмң„м”© м •м§Ғн•ҳкІҢ лӮҙл Өк°‘лӢҲлӢӨ.
- **лӢЁм җ**: ліҙнҸӯ(н•ҷмҠөлҘ )мқҙ кі м •лҗҳм–ҙ мһҲм–ҙм„ң ліөмһЎн•ң лҚ°мқҙн„°м—җм„ңлҠ” мқҙлҰ¬м ҖлҰ¬ нҠҖкұ°лӮҳ, мӣҖн‘№ нҢҢмқё нҠ№м • мөңмҶҢк°’(Local Minima)мңјлЎң кёёмқ„ мһғлҠ” кІҪмҡ°к°Җ л°ңмғқ. мҶҚлҸ„лҸ„ лҠҗлҰј.

### 2) Adam (Adaptive Moment Estimation)

- **нҠ№м§•**: нҳ„мһ¬ л”Ҙлҹ¬лӢқкі„мқҳ м••лҸ„м Ғмқё л°©мӢқ. кҙҖм„ұ(Momentum)кіј м Ғмқ‘м Ғ н•ҷмҠөлҘ (Adaptive, кІҪмӮ¬к°Җ мҷ„л§Ңн•ҳл©ҙ ліҙнҸӯмқ„ нҒ¬кІҢ, к°ҖнҢҢлҘҙл©ҙ ліҙнҸӯмқ„ мһ‘кІҢ мЎ°м Ҳ) л‘җ к°Җм§Җ кі кёү кё°мҲ мқ„ лӘЁл‘җ н•©м№ң м•Ңкі лҰ¬мҰҳ.
- **мһҘм җ**: л°©н–Ҙмқ„ мһғм§Җ м•Ҡкі  м•„мЈј л¶Җл“ңлҹҪкі  л№ лҘҙкІҢ мөңм Ғм җмқ„ м°ҫм•„к°‘лӢҲлӢӨ. мӢӨл¬ҙм—җм„ңлҠ” л¬ҙмЎ°кұҙ `optimizer='adam'`мқ„ кё°ліёк°’мңјлЎң к№”кі  мӢңмһ‘н•©лӢҲлӢӨ.

## 1. IRIS лҚ°мқҙн„° мӢӨмҠө

```python
# мӢӨмҠө 1-0 лҚ°мқҙн„° л¶Ҳлҹ¬мҳӨкё°

iris = load_iris()

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, stratify=iris.target,random_state=260706)

scaler = StandardScaler()
x_train_zs = scaler.fit_transform(x_train)
x_test_zs = scaler.transform(x_test)
```

```python
# мӢӨмҠө 1-1 мӢ кІҪл§қ л¶„м„қ

iris_nn = MLPClassifier((10,50,10), max_iter = 500)
iris_nn.fit(x_train_zs, y_train)
pred = iris_nn.predict(x_test_zs)


print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```

```python
# мӢӨмҠө 1-2 мҳҲмёЎн•  нҷ•лҘ 

iris_nn.predict_proba(x_test_zs)  # 0лІҲмқј нҷ•лҘ , 1лІҲмқј нҷ•лҘ , 2лІҲмқј нҷ•лҘ 
```

## 2. Breast Cancer мӢӨмҠө

```python
# мӢӨмҠө 2-0 лҚ°мқҙн„° л¶Ҳлҹ¬мҳӨкё°

bc = load_breast_cancer()

x_train, x_test, y_train, y_test = train_test_split(bc.data, bc.target, stratify=bc.target,random_state=260706)

scaler = StandardScaler()
x_train_zs = scaler.fit_transform(x_train)
x_test_zs = scaler.transform(x_test)
```

```python
# мӢӨмҠө 2-1 мӢ кІҪл§қ л¶„м„қ

bc_nn = MLPClassifier((64,128,64))
bc_nn.fit(x_train_zs, y_train)
pred = bc_nn.predict(x_test_zs)


print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```

```python
# мӢӨмҠө 2-2 мҳҲмёЎн•  нҷ•лҘ 
bc_nn.predict_proba(x_test_zs) # 0лІҲмқј нҷ•лҘ , 1лІҲмқј нҷ•лҘ 
```

## 4. Dense л Ҳмқҙм–ҙ // FCL

```python
# мӢӨмҠө 4-1 MLPRegressor мһ¬нҳ„

model = Sequential()
model.add(Input(shape=(10,)))              # лҚ°мқҙн„° shape // лҚ°мқҙн„° мһ…кө¬
model.add(Dense(16, activation='sigmoid')) # 16к°ң л…ёл“ң / 0,1 к°’мқё мӢңк·ёлӘЁмқҙл“ңлЎң ліҖнҷҳ
model.add(Dense(8, activation='sigmoid'))  # 16к°ң л…ёл“ң / 0,1 к°’мқё мӢңк·ёлӘЁмқҙл“ңлЎң ліҖнҷҳ
model.add(Dense(1))                        # мөңмў… кІ°кіјк°’
model.summary()

# Param = (мһ…л Ҙ л…ёл“ң мҲҳ * нҳ„мһ¬ мёө л…ёл“ң мҲҳ) + нҳ„мһ¬ мёө л…ёл“ң мҲҳ

# MLPRegressor((16,8)) кіј лҸҷмқј
                   
model.compile(loss='mse',            # мҳӨм°ЁлҘј кі„мӮ°н•ҳлҠ” кё°мӨҖ 
              optimizer='sgd',       # мҳӨм°ЁлҘј мӨ„мқҙлҠ” л°©лІ• / Stochastic Gradient Descent(нҷ•лҘ м Ғ кІҪмӮ¬ н•ҳк°•лІ•)
              metrics=['mae','mse']) # лӢөмқ„ ліҙм—¬мЈјлҠ” л°©мӢқ

hist =model.fit(x_train_zs, y_train, 
          validation_data=(x_test_zs, y_test), # кІҖмҰқ лҚ°мқҙн„°(Validation set)лҘј м§Җм • // м—җнҸӯ(Epoch)мқҙ н•ҳлӮҳм”© лҒқлӮ  л•Ңл§ҲлӢӨ кІҖмҰқ лҚ°мқҙн„°мҷҖмқҳ мҳӨм°Ё кі„мӮ°
          epochs=100)

plt.plot(hist.history['loss'], label='train')
plt.plot(hist.history['val_loss'], label='test')
plt.legend(loc='best')
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.show()
```

```python
pred = model.predict(x_test_zs)
print(pred[:5])
print(y_test[:5])

r2 = r2_score(y_test, pred)
print('R2 score:', r2)
```

## RВІ ScoreлһҖ л¬ҙм—Үмқёк°Җ?

лӘЁлҚёмқҙ лҚ°мқҙн„°лҘј м–јл§ҲлӮҳ лҳ‘лҳ‘н•ҳкІҢ мҳҲмёЎн•ҳлҠ”м§Җ 0м—җм„ң 1 мӮ¬мқҙмқҳ м җмҲҳлЎң нҷҳмӮ°н•ң кІғ.

**мҲҳмӢқ к°ңл…җ**:

$$R^2 = 1 - \frac{лӘЁлҚёмқҳ мҳӨм°Ё}{нҸүк· мңјлЎң мҳҲмёЎн–Ҳмқ„ л•Ңмқҳ мҳӨм°Ё}$$

---

## RВІ Score мқҪлҠ” лІ•

- **RВІ = 1.0 (мөңкі м җ)**: мҷ„лІҪ. лӘЁлҚёмқҙ лҚ°мқҙн„°мқҳ лӘЁл“  ліҖлҸҷмқ„ м„ӨлӘ…. мӢӨл¬ҙм—җм„  м җмҲҳк°Җ мҳӨнһҲл Ө л„Ҳл¬ҙ лҶ’м•„м„ң 'кіјм Ғн•©'мқ„ мқҳмӢ¬н•ҙлҙҗм•ј н•Ё.
- **RВІ = 0.0 (кё°мӨҖм җ)**: м°Ём•…. лӘЁлҚёмқҙ лҚ°мқҙн„°лҘј н•ҳлӮҳлҸ„ м„ӨлӘ…н•ҳм§Җ лӘ»н•Ё.
- **RВІ < 0 (мқҢмҲҳ)**: мөңм•…. лӘЁлҚёмқҙ лӢЁмҲңнһҲ нҸүк· к°’мңјлЎң мҳҲмёЎн•ҳлҠ” кІғліҙлӢӨлҸ„ лҚ” лӘ»н•ң мҲҳм№ҳ. лӘЁлҚё кө¬мЎ°лӮҳ лҚ°мқҙн„° мІҳлҰ¬м—җ мӢ¬к°Ғн•ң л¬ём ң мһҲмқ„ кІғ.
- **мқјл°ҳм Ғмқё м җмҲҳ**: л¶„м•јл§ҲлӢӨ лӢӨлҘҙм§Җл§Ң, 0.7~0.8 мқҙмғҒмқҙл©ҙ "лҚ°мқҙн„°мқҳ нҢЁн„ҙмқ„ кҪӨ мһҳ м°ҫм•„лӮҙкі  мһҲкө¬лӮҳ"лқјкі  нҸүк°Җ.

## 5. BC лҚ°мқҙн„°лЎң Dense л Ҳмқҙм–ҙ м Ғмҡ©

```python
# мӢӨмҠө 5-1 BC лҚ°мқҙн„°лЎң Dense LayerлҘј мқҙмҡ©н•ң мӢ кІҪл§қ лӘЁлҚёл§Ғ

bc = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(bc.data, bc.target, test_size=0.2, random_state=260706)
scaler = StandardScaler()
x_train_zs = scaler.fit_transform(x_train)
x_test_zs = scaler.transform(x_test)

# мӢ кІҪл§қ лӘЁлҚё м„Өкі„
model = Sequential()
model.add(Input(shape=(30,)))             # мҶҚм„ұмқҙ 30к°ң
model.add(Dense(16, activation='relu'))   # мқҖлӢүмёө
model.add(Dense(8, activation='relu'))    # мқҖлӢүмёө
model.add(Dense(1, activation='sigmoid')) # мқҙм§„ л¶„лҘҳмқҙлҜҖлЎң кІ°кіјк°’мқҖ 0~1 мӮ¬мқҙ нҷ•лҘ лЎң лӮҙліҙлӮҙлҠ” sigmoid
model.summary()

'''
л…ёл“ңмқҳ кіјм Ғн•©мқҙлһҖ?

н•ҷмҠөн•  л•Ң лӘЁлҚёмқҙ нӣҲл Ё лҚ°мқҙн„°м—җ л„Ҳл¬ҙ мһҳ л§һм¶°м ём„ң мғҲлЎңмҡҙ лҚ°мқҙн„°м—җлҠ” мһҳ мһ‘лҸҷн•ҳм§Җ м•ҠлҠ” мғҒнғң
мҰү, м¶ңм ңлҗң лӘЁл“  кё°м¶ңл¬ём ңмқҳ лӢөмқ„ мҷёмӣҢм„ң мӢңн—ҳмқ„ м№ҳлҘҙлҠ” кІғкіј к°ҷмқҢ

'''
# лӘЁлҚё н•ҷмҠө к·ңм№ҷ
model.compile(loss='binary_crossentropy', # мқҙм§„ л¶„лҘҳ мҳӨм°Ё кі„мӮ°лІ•
              optimizer='adam',           # sgdліҙлӢӨ м„ұлҠҘмқҙ мўӢмқҢ
              metrics=['accuracy'])       # м •нҷ•лҸ„ ліҙкё°

# н•ҷмҠө м§„н–ү
hist = model.fit(x_train_zs, y_train, 
                 validation_data=(x_test_zs, y_test), # лӘЁмқҳкі мӮ¬ лҚ°мқҙн„°
                 epochs=100)
```

```python
# мӢӨмҠө 5-2 к·ёлһҳн”„ к·ёлҰ¬кё°

plt.figure(figsize=(12, 5)) # к·ёлһҳн”„ нӢҖ

# мҳӨм°Ё(Loss) к·ёлһҳн”„ - мӨ„м–ҙл“ӨмҲҳлЎқ мўӢмқҢ
plt.subplot(1, 2, 1)
plt.plot(hist.history['loss'], label='train loss')
plt.plot(hist.history['val_loss'], label='val loss')
plt.legend()
plt.grid()
plt.title('Loss (Binary Crossentropy)')
plt.xlabel('Epoch')

# м •нҷ•лҸ„(Accuracy) к·ёлһҳн”„ - мҳ¬лқјк°ҲмҲҳлЎқ мўӢмқҢ
plt.subplot(1, 2, 2)
plt.plot(hist.history['accuracy'], label='train acc')
plt.plot(hist.history['val_accuracy'], label='val acc')
plt.legend()
plt.grid()
plt.title('Accuracy')
plt.xlabel('Epoch')

plt.show()
```
