```python
# 실습 #1 아이리스 데이터를 활용한 실습

from sklearn.datasets import load_iris

iris = load_iris()
print(iris.DESCR)
```

```python
iris.data[:10]
```

```python
# 전처리(표준화)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(iris.data)                    # iris 데이터를 활용해서 기준 잡기
iris_data = scaler.transform(iris.data)  # 그 기준을 통해 ()데이터를 조정하기

# 기준을 잡을 데이터의 경우 scaler.fit_transform() 로 한번에 적용가능함


print(iris_data[:10])
```

```python
iris.target[:3]
```

```python
iris_data[:3]
```

```python
# train, test 데이터 분류

from sklearn.model_selection import train_test_split

#   iris_data       iris.target
x_train, x_test, y_train, y_test = train_test_split(iris_data, iris.target,
                                                    test_size=0.3,              # train, test 데이터 규모 조정
                                                    stratify = iris.target,     # 층화추출 적용
                                                    random_state=629)           # 데이터 고정
print(x_train.shape, y_test.shape)
```

```python
# 라벨당 데이터 확인

import numpy as np

print(np.unique(y_test, return_counts=True))  # 불균형 상태 점검
```

```python
# KNN 알고리즘

from sklearn.neighbors import KNeighborsClassifier

iris_knn = KNeighborsClassifier()                 # n_neighbors 값이 K 값(기본값은 5)
iris_knn.fit(x_train, y_train)                    # 학습시키기(데이터 <-> 정답)
pred = iris_knn.predict(x_test)                   # 예측해보기
print(pred)
```

```python
# 예측 검증하기

from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, pred))                          # 정확도(정답 <-> 예측값)




'''
0번을 0으로 15개 예측
1번을 1번으로 14개, 2번으로 1개
2번을 1번으로 2개, 2번으로 13개 분류함
'''
print(confusion_matrix(y_test, pred))                        # 분포도(정답 <-> 예측값)

```

```python
# K값 찾기

for k in range(1,16): # k값을 15까지 돌리기
    iris_knn = KNeighborsClassifier(n_neighbors=k)  # 원래의 k 값은 데이터 수의 제곱근으로 하면 됨(100여개 -> 10여개로)
    iris_knn.fit(x_train, y_train)
    pred = iris_knn.predict(x_test)
    print(k, accuracy_score(y_test, pred))          # 최적의 k값 찾기
```

```python
# 실습 2 train, test 데이터 적용 다른 방식(train 데이터만 표준화 시킨 경우)


x_train, x_test, y_train, y_test = train_test_split(iris_data, iris.target,
                                                    test_size=0.3,              # train, test 데이터 규모 조정
                                                    stratify = iris.target,     # 층화추출 적용
                                                    random_state=629)           # 데이터 고정

# 데이터의 양이 많을 경우(예측할 데이터가 많을 경우) 학습시키는 방식(실무에서 적용됨)

scaler = StandardScaler()
scaler.fit(x_train)                       # train 데이터만 표준화 하여 학습
x_train_std = scaler.transform(x_train)   # train 데이터 조정
x_test_std = scaler.transform(x_test)     # test 데이터 조정
```

```python
# 실습 2-1 K 값 도출

for k in range(1,16):
    iris_knn = KNeighborsClassifier(n_neighbors=k)
    iris_knn.fit(x_train_std, y_train)
    pred = iris_knn.predict(x_test_std)
    print(k, accuracy_score(y_test, pred))
```

```python
# 실습 3 train, test 데이터 적용 다른 방식(정규화 적용)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris.data)
iris_data = scaler.transform(iris.data)

x_train, x_test, y_train, y_test = train_test_split(iris_data, iris.target,
                                                    test_size=0.3,
                                                    stratify = iris.target,
                                                    random_state=629)

for k in range(1,31):
    iris_knn = KNeighborsClassifier(n_neighbors=k)
    iris_knn.fit(x_train, y_train)
    pred = iris_knn.predict(x_test)
    print(k, accuracy_score(y_test, pred))
```

```python
# 실습 4 breast cancer 적용(정규화 vs 표준화)

from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()
print(bc.DESCR)
```

```python
# 실습 4-1 정규화

scaler = MinMaxScaler()
scaler.fit(bc.data)
bc_data = scaler.transform(bc.data)

x_train, x_test, y_train, y_test = train_test_split(bc_data, bc.target,
                                                    test_size=0.3,
                                                    stratify = bc.target,
                                                    random_state=629)

for k in range(1,31):
    bc_knn = KNeighborsClassifier(n_neighbors=k)
    bc_knn.fit(x_train, y_train)
    pred = bc_knn.predict(x_test)
    print(k, accuracy_score(y_test, pred))     # k= 15 에서 98.246%
```

```python
# 실습 4-2 표준화

scaler = StandardScaler()
scaler.fit(bc.data)
bc_data = scaler.transform(bc.data)

x_train, x_test, y_train, y_test = train_test_split(bc_data, bc.target,
                                                    test_size=0.3,
                                                    stratify = bc.target,
                                                    random_state=629)


for k in range(1,31):
    bc_knn = KNeighborsClassifier(n_neighbors=k)
    bc_knn.fit(x_train, y_train)
    pred = bc_knn.predict(x_test)
    print(k, accuracy_score(y_test, pred))            # k=21 에서 98.246%
```

```python
# 실습 5 교차검증 : StratifiedKFold 이해

from sklearn.model_selection import StratifiedKFold

s_kfold = StratifiedKFold(5) # 5조각으로 분류할 것이다 -> 인덱스로 리턴함

                                       # 일반 데이터, target 데이터
for train_idx, test_idx in s_kfold.split(range(100), [0]*50+[1]*50):

# 100개를 5개로 나눠서 20개는 테스트 데이터, 80개는 트레이닝 데이터로 만들어줌
# 일반 데이터 수와 target 데이터 수가 다르면 오류 발생

  print("트레이닝 데이터 \n" , train_idx)
  print("테스트 데이터 \n", test_idx)
```

```python
# 실습 5-1 적용
s_kfold = StratifiedKFold(5)


for train_idx, test_idx in s_kfold.split(bc.data, bc.target):
  scaler = StandardScaler() # 표준화
  scaler.fit(bc.data[train_idx])                 # StratifiedKFold로 분류한 부분에 대해서 train 데이터를 기준으로 적용(스케일 조정)
  x_train = scaler.transform(bc.data[train_idx]) # 해당 기준에 맞춰 train 데이터 적용
  x_test = scaler.transform(bc.data[test_idx])   # 해당 기준에 맞춰 test 데이터 적용

  for k in range(1,50,2):
    bc_knn = KNeighborsClassifier(n_neighbors=k)
    bc_knn.fit(x_train, bc.target[train_idx])
    pred = bc_knn.predict(x_test)
    print(k, accuracy_score( bc.target[test_idx], pred))
  print("========")
```

```python
# 실습 5-2 가장 높은 평균 정확도의 K 값 구하기(나의 풀이)
# 한번에 해주는 코드(자동화) 존재하기에 원리정도만 알기

s_kfold = StratifiedKFold(5)
k_scores = {k: [] for k in range(1, 50, 2)} # K 값을 얼추 알고 있기 때문에 그냥 K 값에 따른 공간을 만들어줌

for train_idx, test_idx in s_kfold.split(bc.data, bc.target):
  scaler = StandardScaler()
  scaler.fit(bc.data[train_idx])
  x_train = scaler.transform(bc.data[train_idx])
  x_test = scaler.transform(bc.data[test_idx])

  for k in range(1,50,2):
    bc_knn = KNeighborsClassifier(n_neighbors=k)
    bc_knn.fit(x_train, bc.target[train_idx])
    pred = bc_knn.predict(x_test)

    score = accuracy_score(bc.target[test_idx], pred) # 각 예측 한 값의 정확도
    k_scores[k].append(score)                         # K_scores에 넣기(딕셔너리 키-벨류 활용)

best_k = 0
best_score = 0

print(" 5-Fold K값 평균 정확도 ")
for k, scores in k_scores.items():
    mean_score = np.mean(scores)
    print(f"K = {k} : {mean_score:.4f}")

    if mean_score > best_score:
        best_score = mean_score
        best_k = k

print("================================")
print(f"최종 : K = {best_k} (평균 정확도: {best_score:.4f})")
```
