```python
# 실습 2-2-2  # 유저의 숫자 입력
while True :                                 # 그냥 계속 반복(break로 중단)
  usernum1 = input()                          # 숫자입력 표시 안나게 만듦
  try :                                      #1 아래 조건이 아니면 오류가 날 수 있는데. 일단 돌려봐
    usernum = list(map(int, usernum1))        # map(a,b) : a를 각각 b의 요소에 적용시켜줌
  except :                                   #1 에러가 발생하면
    print("숫자를 입력하세요.")
    continue                                   # 다시 돌리고 숫자만 입력하라고 안내
  if len(usernum) == 4 :                     #1 입력한 숫자가 4자리라면 진행할거야.
    check = 0                                  #2 중복 체크를 위한 숫자 세팅
    for num in usernum :
      if usernum.count(num) > 1 :              #2 중복된 숫자가 있는지 확인
        check = 1                                 # 같은 숫자가 2이상 있으면 1로 표기
      else :                                   #2 없으면 넘어가
        pass
    if check == 1 :                            #2 check 값이 1이면(-> 중복 존재)
      print("중복되지 않은 4자리를 입력하세요.")
      continue                                   # 안내하고 다시 처음부터 돌려
    else :                                     #2 1)숫자 2)4자리 3)중복없음 조건들을 충족 시 while 반복문 종료
      break
  else :                                     #1 4자리 숫자가 아니면
    print("4자리를 입력하세요.")                 # 안내하고 처음부터 다시 돌려
    continue
print(usernum)
```

```python
# 실습 2-3 # 통합

import random as rd
def generate_randnum() :
  sblst = []
  while len(sblst) < 4:
    tmp = rd.randint(0,9)
    if tmp in sblst :
      continue
    else :
      sblst.append(tmp)
  return sblst
print(generate_randnum())

def user_inputnum() :
  while True :
    usernum = input()
    try :
      usernum = list(map(int, usernum))
    except :
      print("숫자를 입력하세요.")
      continue
    if len(usernum) == 4 :
      check = 0
      for num in usernum :
        if usernum.count(num) > 1 :
          check = 1
      if check == 1 :
        print("중복되지 않은 4자리를 입력하세요.")
        continue
      else :
        break
    else :
      print("4자리를 입력하세요.")
      continue
  return usernum
print(user_inputnum())

def check_result(sblst, usernum):
  strike = 0
  ball = 0
  for idx, num in enumerate(usernum) :
    if num in sblst :
      if usernum[idx] == sblst[idx] :
        strike += 1
      else :
        ball += 1
  return strike, ball

```

```python
# 풀이 2-3 # 게임시작
def play_game() :                                 # 게임시작 함수로 만들었음
  print("Game Start!!")                           # 시작화면은 "Game Start!!"
  sblst = generate_randnum()                      # 무작위 숫자 만들기 함수를 sblst라고 쓸거야.
  print(sblst)                                      # 정답 표시 설정
  while True:                                     # 종료되기 전까지 계속 진행할거야.
    usernum = user_inputnum()                       # 사용자 입력 함수를 usernum이라고 할거야
    strike, ball = check_result(sblst, usernum)     # strike, ball 은 check_result 함수로 돌릴거야.
    if strike == 4 :                                  # 만약에 strike가 4가 나오면
      print("Win")                                     # Win 출력하고
      break                                            # 종료해
    else :                                            # 틀리면 strike와  ball 값 표기해줘
      print("S : ", strike, "B : ", ball)          # 함수이지만 return 을 안써도 되는 이유는 결과물을 보여주지 않아도, 진행이 되기 때문
```
