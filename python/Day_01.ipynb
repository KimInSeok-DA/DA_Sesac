```python
lst1 = list(range(100))
lst_e = []
lst_o = []
# lst1의 요소를 짝수는 lst_e 에, 홀수는 lst_o 에 옮겨놓으시오

# 풀이 #1-2
lst_e = [x for x in lst1 if x % 2 == 0]   # lst_e 에 x를 넣을 건데 [for x] x란 [in lst1] lst1에서 가져다 쓸거야. [if x % 2 == 0] 근데 짝수면 넣어
lst_o = [x for x in lst1 if x % 2 != 0]   # lst_o 에 x를 넣을 건데 [for x] x란 [in lst1] lst1에서 가져다 쓸거야. [if x % 2 != 0] 근데 홀수면 넣어
print(lst_e)
print(lst_o)
print(lst1)
```

```python
# 풀이 #1-3
lst1 = list(range(100))
lst_e = []
lst_o = []

for _ in range(len(lst1)) :     # _라는 작업을 할건데. lst1의 길이를 데이터로 쓸거야
  x = lst1.pop(0)               # lst1의 첫번째 자리를 x라고 하자
  if x % 2 == 0 :               # x가 짝수면
      lst_e.append(x)           # lst_e에 넣어
  else :                        # 그 외(홀수면)
    lst_o.append(x)             # lst_o에 넣어

print(lst1)
print(lst_e)
print(lst_o)
```

```python
#풀이 #2-1
lst1 = list(range(100))
lst_e = []
lst_o = []

while len(lst1)>0 :      # lst1의 길이가 0보다 작거나 같을 때까지 작업할거야
  x = lst1.pop(0)        # lst1의 첫번째 자리를 x 라고 하자
  if x % 2 == 0 :        # x 가 짝수면
    lst_e.append(x)      # lst_e에 넣어
  else :                 # x가 홀수면
    lst_o.append(x)      # lst_o 에 넣어

print(lst1)
print(lst_e)
print(lst_o)
```

```python
#풀이 #2-2
lst1 = list(range(100))
lst_e = []
lst_o = []

while lst1 :          # lst1의 데이터가 없어질 때까지 작업할거야
  x = lst1.pop(0)     # 위와 동일
  if x % 2 == 0 :
    lst_e.append(x)
  else :
    lst_o.append(x)

print(lst1)
print(lst_e)
print(lst_o)
```

```python
## 학생의 점수 평균 내기 ##
scores = [[70,80,90],[50,40,20],[70,50,60]]
print(scores)
```

```python
# 학생의 점수들 도출
scores[0]
```

```python
# 세부 과목 도출
scores[0][0]
```

```python
#내 풀이
a = scores[0]
b = scores[1]
c = scores[2]
a_avr = sum(a) / len(a)
b_avr = sum(b) / len(b)
c_avr = sum(c) / len(c)
print(a_avr)
print(b_avr)
print(c_avr)
```

```python
# 풀이 #3-1
scores = [[70, 80, 90], [50, 40, 20], [70, 50, 60]]

for score in scores:                      # score 작업을 할건데. 이건 scores 에서 가져다 써
    avg = sum(score) / len(score)         # avg 는 score 합계 / score의 길이 인거야
    print(f"{avg:.2f}")                   # 출력해                                                               f"{aaaa:.2f}" aaaa를 소수점 둘째자리까지 표시
```

```python
# 풀이 #3-2
a_score = scores[0]                          # a_score는 scores의 첫번째 데이터야
b_score = scores[1]
c_score = scores[2]

print(f"{sum(a_score) / len(a_score):.2f}")  # a_score를 다 더하고, 길이(개수)로 나눠. 그리고 소수 둘째자리까지 출력해
print(f"{sum(b_score) / len(b_score):.2f}")
print(f"{sum(c_score) / len(c_score):.2f}")
```

```python
# 풀이 #3-3
scores = [[70,80,90],[50,40,20],[70,50,60]]

for score1 in scores:           # 1 작업을 할거야. 이건 scores에서 데이터를 가져다가 써 scores 다 할 때까지해
  total = 0                     # 처음 합계는 0이야.
  for score2 in score1 :        # 2 작업을 할건데. 이건 1에서 써
    total += score2             # 합계에서 score2 에 있는 걸 더하고 다하면 다음 score1로 진행해
  print("mean :",total/3)
```

```python
# 풀이 #3-4
for st in scores :                    # st 작업을 할거야. 이건 scores에서 가져다가 써
  print("평균 : ",  sum(st)/len(st))   # 출력할 건데 평균: 을 먼저 출력하고, st 합/길이 를 계산해서 붙여
```

```python
# 풀이 #3-5
st_mean = [f"{sum(st)/len(st):.2f}" for st in scores]   # st_mean 이란 건데. 이건 st의 합을 길이로 나눌건데. 소수점 둘째자리까지만 표시해. 그리고 st는 scores에서 가져다가 써
print(st_mean)                                          # st_mean 결과는?
```

```python
# 정렬
sorted(st_mean) # st_mean.sort 는 값이 변경되버림
```
