### 미션 1. 명언 사이트 크롤링

- 수집 URL: `https://quotes.toscrape.com/`
- 수집 목표
  - 명언 문장
  - 작성자

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://quotes.toscrape.com/"                # URL 불러오기
response = requests.get(url)                        # 연결하기
soup = BeautifulSoup(response.text, "html.parser")  # 작업 준비하기

print("성공")
```

```python
# div에서 "quote" 인것 찾아서 넣기

quote_boxes = soup.find_all("div", class_="quote")  

print(len(quote_boxes))
```

```python
# 내용 추출해서 records에 넣기

records = []  

for box in quote_boxes :
    quote_txt = box.find("span", class_="text").get_text(strip=True)
    author = box.find("small", class_="author").get_text(strip=True)
    records.append([quote_txt, author])

records[:3]
```

```python
# pandas로 데이터 프레임 만들기

df_quotes = pd.DataFrame(records, columns=["명언", "작성자"])

df_quotes.head()
```
