# 0. 환경설정

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import cluster
from sklearn.decomposition import PCA
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
```

# 1. K-Means 실습

```python
# 실습 1-0 K-means 실습
# [X좌표, Y좌표], [정답]
data = make_moons(10,             # 데이터 수
                  noise = 0.05,     # 무작위 오차 설정
                  random_state =3)  # 시드값 고정
print(data)
```

```python
# 실습 1-1 K-means 실습

data = make_moons(1000,             # 데이터 수
                  noise = 0.05,     # 무작위 오차 설정
                  random_state =3)  # 시드값 고정
#print(data)
               # 라벨
plt.scatter(data[0][:,0], data[0][:,1]) # 첫번째 데이터 가져오기
plt.show()
```

```python
# 실습 1-2 전처리 후

scaler = StandardScaler()
x_data = scaler.fit_transform(data[0])
plt.scatter(x_data[:,0], x_data[:,1])
plt.show()
```

```python
# KMeans(

# n_clusters: int = 8,                  # 군집 개수
# init: str = "k-means++",              # 초기 중심점 선택(기본값은 서로 멀리 설정한다라는 의미)
# n_init: str = "auto",                 # 재시도 횟수
# max_iter: int = 300,                  # 최대 반복 횟수
# tol: float = 0.0001,                  # 수렴기준
# verbose: int = 0,                     # 진행도 확인
# random_state: Unknown | None = None,  # 시드값
# copy_x: bool = True,                  # 데이터 복사여부(원본 건드는지)
# algorithm: str = "lloyd"              # 알고리즘 종류
```

```python
# 실습 1-3 KMeans # 거리 기반 군집

cluster = KMeans(n_clusters=2)
pred = cluster.fit_predict(x_data)
plt.scatter(x_data[:,0], x_data[:,1], c=pred)
plt.show()
```

```python
# DBSCAN(

# * eps: float = 0.5,                    # 군집 탐색 범위
# * min_samples: int = 5,                # 군집이 이루어지는 최소 범위
# * metric: str = "euclidean",           # 거리 측정 방식
# metric_params: Unknown | None = None,# 거리측정방식에 추가적인 옵션
# * algorithm: str = "auto",             # 계산 알고리즘
# leaf_size: int = 30,                 # 거리계산을 빠르게 하는 데이터 구조 크기 설정
# p: Unknown | None = None,            # 'minkowski'사용시 적용(1 이면 맨해튼, 2이면 유클리드)
# n_jobs: Unknown | None = None        # 학습에 사용하는 cpu 개수
```

```python
# 실습 1-4 DBSCAN # 밀도 기반 군집

cluster = DBSCAN()
pred = cluster.fit_predict(x_data)
plt.scatter(x_data[:,0], x_data[:,1], c=pred)
plt.show()
```

# 2. Iris 데이터를 활용한 적용

```python
# 실습 2-4 주성분분석(Principal Component Analysis) 적용
# 차원축소 기법 중 주성분분석


pca = PCA()
iris_pca = pca.fit_transform(x_data)
print(iris_pca)
plt.scatter(iris_pca[:,0], iris_pca[:,1], c=iris.target)
plt.show()
```

# 3. 연관분석

```python
# class 가 아니고 함수임

# apriori(

# df: Unknown,                    # 데이터 프레임
# min_support: float = 0.5,       # 최소 지지도 설정
# use_colnames: bool = False,     # 항목 집합에 컬럼 표시 여부
# max_len: Unknown | None = None, # 집합의 최대 길이 개수 설정
# verbose: int = 0,               # 학습과정 로그 출력
# low_memory: bool = False        # 메모리 사용량 최적화 여부 설정
# )
```

```python
# class 가 아니고 함수임

# association_rules(

# df: DataFrame,                            # 데이터 프레임
# num_itemsets: int | None = 1,             # df 가 어떤 결과물인가 구분
# df_orig: DataFrame | None = None,         # 원본 데이터 참조할 때 사용
# null_values: bool = False,                # 결측치를 null로 처리할지
# metric: str = "confidence",               # 기준이 되는 지표
# min_threshold: float = 0.8,               # 최소 기준값
# support_only: bool = False,               # 지지도 외 계산 생략 여부
# return_metrics: list[Unknown] = _metrics  # 데이터 프레임에 어떤 지표 포함하는지 지정
```

```python
# 실습 3-1 연관분석

dataset = [
    ['우유', '기저귀','맥주' ],
    ['우유', '기저귀'],
    ['맥주', '기저귀'],
    ['우유', '빵'],
    ['맥주', '빵']
]

te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_data, columns=te.columns_)
print(df.head())

freq_itms = apriori(df,
                    min_support=0.2,   # support 가 20% 이상 나오는 것만
                    use_colnames=True)
print(freq_itms)

rules = association_rules(freq_itms, metric='lift', min_threshold=1.0)
print(rules[['antecedents', # 선행항목
             'consequents', # 후행항목
             'support',     # 지지도
             'confidence',  # 신뢰도
             'lift'         # 향상도
             ]])
```

# 4. 타이타닉

```python
# 실습 4-0 데이터 준비

raw_data = sns.load_dataset('titanic')
raw_data.head()
```

```python
# 실습 4-1 생존 관련 데이터 보기

df = raw_data[['survived', 'class', 'sex', 'age']].dropna().copy()
df.head()
```

```python
# 실습 4-2 survived 조정

df['survived'] = df['survived'].map({0:'사망', 1:'생존'})
df.head()
```

```python
# 실습 4-3 연령 조정

bns = [0, 19, 35, 60, 100]
lbls = ['어린이', '청년', '중년', '노년']
df['age'] = pd.cut(df['age'], bins= bns, labels = lbls)

df.head()
```

```python
# 실습 4-4 연관 분석 준비 = one-hot encoding

df_encd = pd.get_dummies(df)
df_encd.head()
```

```python
# 실습 4-5 연관분석

freq_itms = apriori(df_encd,
                    min_support=0.05,   # support 가 5% 이상 나오는 것만 / 어린이 수가 적음
                    use_colnames=True)

rules = association_rules(freq_itms, metric='lift', min_threshold=1.0)
rules.sort_values(by='lift', ascending=False)
```

```python
# 실습 4-6 연관분석 후 일부 데이터만 확인하기

target_rules = rules[rules['consequents'].apply(
    lambda x:any('survived' in item for item in x))]

# confidene : antecedents 가 consequents 하는 확률
# lift : antecedents 가 전체 평균보다 consequents 하는 확률

target_rules.sort_values(by='lift', ascending=False)
```
