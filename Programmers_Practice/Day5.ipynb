# 1. 코드 처리하기

문자열 code가 주어집니다.

code를 앞에서부터 읽으면서 만약 문자가 "1"이면 mode를 바꿉니다. mode에 따라 code를 읽어가면서 문자열 ret을 만들어냅니다.

mode는 0과 1이 있으며, idx를 0 부터 code의 길이 - 1 까지 1씩 키워나가면서 code[idx]의 값에 따라 다음과 같이 행동합니다.

mode가 0일 때

code[idx]가 "1"이 아니면 idx가 짝수일 때만 ret의 맨 뒤에 code[idx]를 추가합니다.

code[idx]가 "1"이면 mode를 0에서 1로 바꿉니다.

mode가 1일 때

code[idx]가 "1"이 아니면 idx가 홀수일 때만 ret의 맨 뒤에 code[idx]를 추가합니다.

code[idx]가 "1"이면 mode를 1에서 0으로 바꿉니다.

문자열 code를 통해 만들어진 문자열 ret를 return 하는 solution 함수를 완성해 주세요.

단, 시작할 때 mode는 0이며, return 하려는 ret가 만약 빈 문자열이라면 대신 "EMPTY"를 return 합니다.

입력값 "abc1abc1abc"

출력값 "acbac"

```python
# 풀이 1
def solution(code):
    ret = ''
    mode = 0                                 # 시작할 때 mode는 0
    for idx, char in enumerate(code):        # idx(인덱스)와 문자를 동시에 가져오기 위해 enumerate를 사용합니다.
        if char == "1":                       #  문자가 "1"이면 mode를 전환. (0 -> 1 또는 1 -> 0)
            if mode == 0:
                mode = 1
            else:
                mode = 0
        else:                                 #  문자가 "1"이 아닐 때 (원래 글자일 때)
            if mode == 0:                      # mode가 0일 때는 idx가 짝수(2로 나눈 나머지가 0)일 때만 추가
                if idx % 2 == 0:
                    ret += char
            else:                              # mode가 1일 때는 idx가 홀수(2로 나눈 나머지가 1)일 때만 추가
                if idx % 2 == 1:
                    ret += char
    if ret == '':                           # 3. 예외 처리: 만약 ret가 빈 문자열이면 "EMPTY"를 반환
        return "EMPTY"

    return ret
```

#2. 등차수열에 특정항 더하기

두 정수 a, d와 길이가 n인 boolean 배열 included가 주어집니다.

첫째항이 a, 공차가 d인 등차수열에서 included[i]가 i + 1항을 의미할 때,

이 등차수열의 1항부터 n항까지 included가 true인 항들만 더한 값을 return 하는 solution 함수를 작성해 주세요.

```python
# 풀이 1

def solution(a, d, included):
  answer = 0
  for idx, is_true in enumerate(included):     # included 배열을 순서대로 돌면서 idx(방 번호)와 진짜값(True/False)을 가져옵니다.
    if is_true:                                # 만약 해당 항이 True 라면?
      c_term = a + (idx * d)                   # 등차수열의 i번째 항의 값 계산 공식: a + (idx * d)
      answer += c_term                         # 누적해서 더해줍니다.
    return answer
```

```python
# 풀이 2

def solution(a, d, included):
  return sum(a + i * d for i, f in enumerate(included) if f)
```

#3. 주사위 게임2

1부터 6까지 숫자가 적힌 주사위가 세 개 있습니다. 세 주사위를 굴렸을 때 나온 숫자를 각각 a, b, c라고 했을 때 얻는 점수는 다음과 같습니다.

세 숫자가 모두 다르다면 a + b + c 점을 얻습니다.

세 숫자 중 어느 두 숫자는 같고 나머지 다른 숫자는 다르다면 (a + b + c) × (a2 + b2 + c2 )점을 얻습니다.

세 숫자가 모두 같다면 (a + b + c) × (a2 + b2 + c2 ) × (a3 + b3 + c3 )점을 얻습니다.

세 정수 a, b, c가 매개변수로 주어질 때, 얻는 점수를 return 하는 solution 함수를 작성해 주세요.

```python
# 풀이 1

def solution(a, b, c):
    if a == b == c:                                                            # 1단계: 세 숫자가 모두 같은 경우
        answer = (a + b + c) * (a**2 + b**2 + c**2) * (a**3 + b**3 + c**3)
    elif a != b and b != c and a != c:                                         # 2단계: 세 숫자가 모두 다른 경우 (어느 두 개도 같지 않음)
        answer = a + b + c
    else:                                                                      # 3단계: 위의 두 조건이 아니라면 '어느 두 숫자만 같은 경우'에 해당합니다.
        answer = (a + b + c) * (a**2 + b**2 + c**2)
    return answer
```

# 4. 원소들의 곱의 합

정수가 담긴 리스트 num_list가 주어질 때, 모든 원소들의 곱이 모든 원소들의 합의 제곱보다 작으면 1을 크면 0을 return하도록 solution 함수를 완성해주세요.

```python
# 풀이 1

def solution(num_list):
    multiply_result = 1
    for num in num_list:                 # 모든 원소들의 곱 구하기
        multiply_result *= num
    sum_squared = sum(num_list) ** 2     # 모든 원소들의 합의 제곱 구하기
    if multiply_result < sum_squared:    # 조건 비교해서 결과 반환하기
        return 1
    else:
        return 0
```

```python
# 풀이 2

def solution(num_list):
    a=1
    b=0
    for x in num_list:
        a*=x
        b+=x
    if a<b*b: return 1
    return 0
```

# 5. 이어붙인 수

정수가 담긴 리스트 num_list가 주어집니다.

 num_list의 홀수만 순서대로 이어 붙인 수와 짝수만 순서대로 이어 붙인 수의 합을 return하도록 solution 함수를 완성해주세요.

```python
# 풀이 1

def solution(num_list):
    odd_str = ""   # 홀수들을 이어 붙일 빈 문자열 주머니
    even_str = ""  # 짝수들을 이어 붙일 빈 문자열 주머니

    for num in num_list:
        if num % 2 == 1:
            odd_str += str(num)
        else:
            even_str += str(num)
    answer = int(odd_str) + int(even_str)
    return answer
```

```python
# 풀이 2
def solution(num_list):
    odd = int("".join([str(x) for x in num_list if x % 2 == 1]))     # 홀수만 뽑아서 문자로 만든 뒤 조인(join)하고 숫자로 바꾸기
    even = int("".join([str(x) for x in num_list if x % 2 == 0]))    # 짝수만 뽑아서 문자로 만든 뒤 조인(join)하고 숫자로 바꾸기
    return odd + even
```
