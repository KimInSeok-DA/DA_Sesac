생각 순서
1. lottos의 번호가 win_num에 있는지 확인(있는 경우, 없는 경우, 0인 경우 정리)
2. 0으로 표기된 부분의 경우 win_num에 전부 있다고 가정하는 최고 순위, 없다고 가정하는 최저 순위
3. 최고 순위는 check_nums+zero_nums
4. 최저 순위는 check_nums만
5. 최고 순위와 최저순위에 따라 등수 표기

```python
# 풀이 1

lottos = [4,0,3,0,2,8]
win_num = [3,45,2,22,12,10]

def solution(lottos, win_num):
  check_nums = 0
  zero_nums = 0
  none_nums = 0
  for num in lottos :                      # lottos에 있는 번호를 맞춰보는 작업
    if num in win_num :                     #1 그 번호가 win_num에 있으면
      check_nums += 1                         #2 check_nums에 1더하기
    elif num == 0 :                         #1 0이라면
      zero_nums +=1                           #2 zero_nums에 1더하기
    else :                                  #1 없으면
      none_nums +=1                           #2 none_nus에 1더하기
  max_rank = 7-(zero_nums+check_nums)      # 순위 -> 맞힌 개수의 역
  min_rank = 7-(check_nums)

  if max_rank > 6:      # ai 돌린 구간
    max_rank = 6
  if min_rank > 6:
    min_rank = 6        # ai 돌린 구간

  lst1 = [max_rank, min_rank]
  return lst1

print(solution(lottos, win_num))
```
