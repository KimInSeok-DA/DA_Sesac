# 1. 문자열 섞기
길이가 같은 두 문자열 str1과 str2가 주어집니다.

두 문자열의 각 문자가 앞에서부터 서로 번갈아가면서 한 번씩 등장하는 문자열을 만들어 return 하는 solution 함수를 완성해 주세요.

str1 "aaaaa"

str2	"bbbbb"

result "ababababab"

```python
# 풀이 1
str1 = 'aaaaa'
str2 = 'bbbbb'
def solution(str1, str2):
    answer = ''
    for s1, s2 in zip(str1, str2):
        answer += s1
        answer += s2
    return answer
```

```python
# 풀이 2
def solution(str1, str2):
    answer = ''
    for i in range(0,len(str1)):
        answer = answer + str1[i] + str2[i]
    return answer
```

# 2. 문자 리스트를 문자열로 변환하기
문자들이 담겨있는 배열 arr가 주어집니다. arr의 원소들을 순서대로 이어 붙인 문자열을 return 하는 solution함수를 작성해 주세요.

arr 입력  ["a","b","c"]
result "abc"

```python
# 풀이 1

def solution(arr):
    answer = ''
    for c in arr:
        answer += c
    return answer
```

```python
# 풀이 2

def solution(arr):
    return ''.join(arr)
```

# 4. 더 크게 합치기

연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다

양의 정수 a와 b가 주어졌을 때, a ⊕ b와 b ⊕ a 중 더 큰 값을 return 하는 solution 함수를 완성해 주세요.

단, a ⊕ b와 b ⊕ a가 같다면 a ⊕ b를 return 합니다.

a 9
b 91
result 991

```python
# 풀이 1
def solution(a, b):
    ab = f"{a}{b}"
    ba = f"{b}{a}"
    ab_num = int(ab)
    ba_num = int(ba)

    if ab_num >= ba_num :
        return ab_num
    else :
        return ba_num
```

```python
# 풀이 2
def solution(a, b):
    return max(int(f"{a}{b}"), int(f"{b}{a}"))
```

# 5. 두 수의 연산값 비교
연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다.

12 ⊕ 3 = 123
3 ⊕ 12 = 312
양의 정수 a와 b가 주어졌을 때, a ⊕ b와 2 * a * b 중 더 큰 값을 return하는 solution 함수를 완성해 주세요.

단, a ⊕ b와 2 * a * b가 같으면 a ⊕ b를 return 합니다.

a 2
b 92
result 364

```python
# 풀이 1
def solution(a, b):
    num = int(f"{a}{b}")
    math = 2 * a * b
    if num >= math:
        return num
    else:
        return math
```

```python
# 풀이 2
def solution(a, b):
    return max(int(str(a) + str(b)), 2 * a * b)
```
