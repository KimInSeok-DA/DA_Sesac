# 굿모닝 코딩 : 크기가 작은 부분 문자열

숫자로 이루어진 문자열 t와 p가 주어질 때,

t에서 p와 길이가 같은 부분문자열 중에서,

이 부분문자열이 나타내는 수가 p가 나타내는 수보다 작거나 같은 것이 나오는 횟수를 return하는 함수 solution을 완성하세요.

예를 들어, t="3141592"이고 p="271" 인 경우,

t의 길이가 3인 부분 문자열은 314, 141, 415, 159, 592입니다.

이 문자열이 나타내는 수 중 271보다 작거나 같은 수는 141, 159 2개 입니다.

생각 순서
1. P의 문자열의 길이 -> len
2. T의 문자열을 len만큼 비교하되, [한칸씩 밀려나게끔] <- 이걸 어찌하오
3. 비교한 것이 P보다 작거나 같으면 answer에 +1
4. answer에 담겨진 개수 도출

```python
# 풀이 이전 range 복습
print(list(range(0,10,2)))    # 0~9까지인데, 2만큼 간격이 있어.
print(list(range(0,10)))      # 한칸씩 띄어야한다면 세번째는 공백으로 유지

t='3141592'                   # -> 314,131,159,592
t1='500220839878'             # -> 500,002, 022, .....
p="271"                       # 3개씩
p1="7"                        # 1개씩
n=len(p)

print(f"한칸씩 간다면 {list(range(0,len(t)-n+1))}")

def solution(t, p):
  answer = []
  n=len(p)
  for i in range(0, len(t) - n + 1): # ai 사용                                                              # i 라는 작업을 할건데. i는 한칸씩 진행하는 것
    if int(t[i:i+n]) <= int(p) :     # 처음엔 일단 했지만 오류가 나서 오류값을 ai 물어봐서 수정함               # t의 [첫번째 자리부터 p의 길이만큼의 값] 한칸씩 진행 하는 값들이 p보다 작거나 같으면
      answer.append(t[i:i+n])                                                                               # 그 값을 answer에 넣음
  return answer
result = solution(t, p)
print(result)
```

```python
# 나의 풀이

t='3141592'
t1='500220839878'
p="271"
p1="7"

def solution(t, p):
  answer = 0                           # 문제에선 개수만 물어봄
  n=len(p)
  for i in range(0, len(t) - n+1):
    if int(t[i:i+n]) <= int(p) :
      answer +=1
  return answer
result = solution(t, p)
print(result)
```
