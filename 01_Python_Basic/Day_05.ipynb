생각 순서
1. lottos의 번호가 win_num에 있는지 확인(있는 경우, 없는 경우, 0인 경우 정리)
2. 0으로 표기된 부분의 경우 win_num에 전부 있다고 가정하는 최고 순위, 없다고 가정하는 최저 순위
3. 최고 순위는 check_nums+zero_nums
4. 최저 순위는 check_nums만
5. 최고 순위와 최저순위에 따라 등수 표기


```python
# 풀이 1

lottos = [4,0,3,0,2,8]
win_num = [3,45,2,22,12,10]

def solution(lottos, win_num):
  check_nums = 0  # 교수님 화면 보고 작성함
  zero_nums = 0
  none_nums = 0
  for num in lottos :                      # lottos에 있는 번호를 맞춰보는 작업
    if num in win_num :                     #1 그 번호가 win_num에 있으면
      check_nums += 1                         #2 check_nums에 1더하기
    elif num == 0 :                         #1 0이라면
      zero_nums +=1                           #2 zero_nums에 1더하기
    else :                                  #1 없으면
      none_nums +=1                           #2 none_nus에 1더하기
  max_rank = 7-(zero_nums+check_nums)      # 순위 -> 맞힌 개수의 역
  min_rank = 7-(check_nums)

  if max_rank > 6:      # ai 돌린 구간
     max_rank = 6
  if min_rank > 6:
    min_rank = 6        # ai 돌린 구간

  lst1 = [max_rank, min_rank]
  return lst1

print(solution(lottos, win_num))
```

```python
# 풀이 2(by 윤창기님)
lottos = [0, 0, 0, 0, 0, 0]
win_nums = [31, 10, 45, 1, 6, 19]

# zero_count + same_number: 맞힌 갯수


def solution1(lottos: list, win_nums: list) -> list:

    dict1 = {6: 1, 5: 2, 4: 3, 3: 4, 2: 5, 1: 6}                          # 6개 맞추면 1등을 딕셔너리로 표현

    zero_count = lottos.count(0)                                          #A 0의 개수
    same_number = set(lottos).intersection(win_nums)                      #B same_number란 lottos의 집합 중 win_nums와 겹치는 것들
    sum = zero_count + len(same_number)                                     # A와 B의 개수 합
    a = 6 if dict1[sum] + zero_count == 7 else dict1[sum] + zero_count    # a는 만약 sum의 값에 해당하는 딕셔너리의 벨류값과 0의 개수의 합이 7이라면 6 이고
                                                                          # 그외에는 벨류값과 0의 개수의 합이야.
    return [dict1[sum], a]                                                  # 최고순위는 sum(맞춘 개수+0의 개수)을 키로 판단한 딕셔너리의 벨류값, 최저는 a


def solution2(lottos: list, win_nums: list) -> list:
    rank = {6: 1, 5: 2, 4: 3, 3: 4, 2: 5}
    zero_count = lottos.count(0)
    match_count = len(set(lottos) & set(win_nums) - {0})               # match_count는 lottos의 집합화 그리고 win_nums의 집합화한 것의 겹치는 것들에서 0인 것들을 제외 한 것의 길이(개수)

    max_match = match_count + zero_count                                # 0이 전부 당첨이라 가정 (최고)
    min_match = match_count                                             # 0이 전부 꽝이라 가정 (최저)

    return [rank.get(max_match, 6), rank.get(min_match, 6)]            # 최고순위의 경우 rank 딕셔너리에서 max_match를 가져오되, 값이 없는 경우(1,0은 낙점인 6등)는 6을 출력
                                                                       # 최저순위의 경우도 동일

print(solution1(lottos, win_nums))
```

```python
# 풀이 3 (by 류경혜님)
def solution(lottos, win_nums):
  same_num = sum(lottos[i] in win_nums for i in range(6))           # lottos와 win_nums 중 같은 숫자의 개수
  count_0 = lottos.count(0)                                         # lottos에 존재하는 0의 개수

  # (1). 6과 계산한 값 중 작은 값을 선택함                 (2) 계산한 값이 7이라면 6으로, 아니라면 계산한 값으로
  max_ranking = min(6, 7 - (same_num + count_0))            # max_ranking= 6 if 7 - (same_num + count_0) == 7 else 7 - (same_num + count_0)
  min_ranking = min(6, 7 - same_num)                        # min_ranking = 6 if 7 - (same_num) == 7 else 7 - (same_num)

  return [max_ranking, min_ranking]

solution([44, 1, 0, 0, 31, 25],	[31, 10, 45, 1, 6, 19]) # [3, 5]
```

```python
# 실습 1-1
class Counter():
  def reset(self):          # reset 이란 기능은
    self.count = 0           # 입력한 항목에 대한 count 값(개수)을 0으로 설정함
  def increment(self):      # increment란 기능은
    self.count +=1           # 입력한 항목에 대한 count 값(개수)를 1 더함
  def get(self):            # get이란 기능은
    return self.count        # 입력한 항목의 count 값을 돌려줌

a = Counter()          # Counter로 만들어진 a 가 있다.
a.reset()               # a에 대한 Counter의 reset 기능 실행        -> a = 1 로 설정
a.increment()           # a에 대한 Counter의 increment 기능 실행    -> a 값에 + 1 하기

print(a.get())          # a에 대한 Counter의 get 기능 실행          -> a 값 출력
```

```python
# 실습 1-4
class Counter():
  def __init__(self):
    self.__count = 0           # __를 붙이면 임의로 값이 변경되지 않게 함 -> 생성자라고 함
  def reset(self):               # 이후 print() 적용시 숨겼기 때문에 반영은 되지 않음
    self.__count = 0
  def increment(self):
    self.__count +=1
  def get(self):
    return self.__count

a = Counter()
a.increment()
print(a.get())
a.count = 10
print(a.get())
```

```python
# 풀이 # 나의 풀이

class Person() :
  def __init__(self, name=None, age = 0):
    self.__name = name
    self.__age = age
  def getName(self):                                                      # 사용자가 입력한 name [불러오기]
    return self.__name
  def getAge(self):                                                       # 사용자가 입력한 age [불러오기]
    return self.__age
  def setName(self, name):
    self.__name = name                                                    # 사용자가 입력한 name 로 [설정]
  def setAge(self, age):
    self.age = age                                                        # 사용자가 입력한 age 로 [설정]
  def __str__(self) :
    return f"Name = {self.__name}, Age = {self.__age}"
  def __eq__(self, other):
    return (self.__name == other.__name) and (self.__age == other.__age)

a = Person()
b = Person("kim", 10)

if a==b:
  print("같음")
else :
  print("다름")

print(a)
```

```python
# 실습 4 list 기능 만들어보기
class mylist():
  def __init__(self, data=[]) :      # 기본값 빈 공간임을 설정
    self.__data = data                 # 사용자가 입력한 정보에 .__data란 내장된 요소 를 의미
  def append(self, item):            # append 기능이란
    self.__data.append(item)           # 사용자 정보에 item을 추가한다
  def __str__(self):                 # str 기능이란
    return f"{self.__data}"            # 입력값을 출력할 때 형태
  def __getitem__(self, idx):        # getitem 기능이란
    return self.__data[idx]            # 사용자 입력값의 위치임
  def __add__(self, rhs):            # add 기능이란
    ret=list(self.__data)              # 사용자 입력값을 ret에 넣고
    for additem in rhs:                  # 추가하는 rhs의 값을
      ret.append(additem)                # ret에 전부 넣음
    return ret                           # 그리고 ret를 반환함

#mlst1=list()
mlst1 = mylist()
mlst1.append(10)
print(mlst1)
print(mlst1[0])

#mlst2 = mylist([3,4,5])
#mlst2=list([3,4,5])
#mlst3 = mlst1 + mlst2
#print(mlst3)
```

```python
# 실습 5 상속
class stu(Person):
  def __init__(self, name=None, age=0, major=None):        # 하위 클래스의 기본값 설정
    super().__init__(name,age)                              # 데이터를 물려받을 요소 설정
    self.__major = major                                    # 하위 클래스 자체 데이터 설정
  def getMajor(self):                                      # getMajor 기능이란
    return self.__major                                     # 사용자가 입력한 값의 major 불러오기
  def setMajor(self, major):                               # setMajor 기능이란
    self.__major = major                                    # 사용자가 입력한 값의 major 설정

kang = stu("kang", 26, "cs")
print(kang.getName())
kang1 = stu("kang", 26, "ds")
print(kang1.getAge())

if kang ==kang1 :                        # Person 에서는 같다고 판단하여 "같음" 결과 나옴
  print("같음")

```

```python
# 실습 5-1 상속 자체에서의 별도 비교
class stu(Person):
  def __init__(self, name=None, age=0, major=None):
    super().__init__(name,age)
    self.__major = major
  def getMajor(self):
    return self.__major
  def setMajor(self, major):
    self.__major = major
  def __eq__(self, other):                                               # 하위 클래스의 조건까지 비교하기(하위 클래스 우선 적용)
    return super().__eq__(other) and (self.__major == other.__major)       # 상위 클래스의 비교데이터(eq) 불러오기 + 추가 조건까지 동등한지 확인

kang = stu("kang", 26, "cs")
print(kang.getName())
kang1 = stu("kang", 26, "ds")
print(kang1.getAge())

if kang ==kang1 :
  print("같음")
else :
  print("다름")
```
