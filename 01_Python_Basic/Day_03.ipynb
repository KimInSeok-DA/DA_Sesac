```python
# 풀이 4-1
def hidenames(members) :               # members 를 통해 익명화하는 hedenames를 함수를 만들거야
  newnames = []                        # newnames 란 리스트를 사용할 거야.
  for name in members :
    hide = name[0] + '*' + name[2]     # hide 는 name의 첫번째 글자 + * + 두번째 글자의 조합이야.  /// 성이 2글자인경우 원치않은 결과 발생
  # hide = name[:-2] + "*" + name[:-1] # 외자의 경우 name[:-2] 없는 걸로 인식되어 *석으로 출력
    newnames.append(hide)              # newnames에 hide를 추가해
  return newnames                      # newnames 로 결과값을 출력해

new_members = hidenames(members)       # new_members 는 members를 hidenames 함수로 돌린 거야.
print(new_members)
```

```python
# 풀이 4-2
def hide_name(members) :
  newnames = []
  for name in members :
    hide = name.replace(name[-2], '*')  # hide는 name의 뒤에서 두번째 글자를 *로 바꿔서 보여주는 거야.
    newnames.append(hide)               # 같은 글자 2회이상 겹치는 거 주의
  return newnames                       ## 관건은 적용했을 경우 예외 발생에 주의해야함

new_members = hidenames(members)
print(*new_members)                     # * : 표기 시 [], '' 삭제 시켜줌
```

```python
# 풀이 6 딕셔너리에 같은 성을 가진 사람 묶기
members = ['김찬은', '송문택', '윤창기', '최진주', '류경혜',
           '김진경', '최호연', '노현동', '배원준', '정선영',
           '민승희', '임지수', '안지예', '김민주', '한예진',
           '유은영', '김인석', '이수인', '정종현', '허찬빈',
           '진혜원', '남궁세정']

# 풀이 6-1
lastname = {}
for a in members :
  b = a[:-2]                  # [:-2] 대신 [0]으로 해도 되지만 성씨가 두글자인 예외를 감안하여 이름이 전부 두글자 기준으로 실행.
  if b in lastname :
    lastname[b].append(a)
  else :
    lastname[b] = [a]

print(lastname)
```

```python
# 풀이 6-2
lastname = {}                                                 # 딕셔너리 : 방 배정 시스템 {방번호 : 사람}
for name in members :                                         # name 작업을 할건데. members 전부 돌릴때까지 해(0부터 시작) * name[:-2] 는 편의상 성씨라고 하겠음
  lastname[name[:-2]] = lastname.get(name[:-2], []) + [name]  # lastname에 성씨 는 어떻게 이뤄지냐면 = 성씨가 있으면 이름을 넣어주고, // 없으면 [](새로운 방)을 새로 만들어서 [name] 이름을 넣어줘
# 저장할 방 주소(키)   = 저장할 내용(벨류값) 조건 + 넣을 내용
print(lastname)    # lastname[name[:-2]] = 에서 name[:-2] 가 없으면 저장할 곳을 잃어버림 "작업은 했는데. 어디다가 넣어요..?" -> 딕셔너리 특징임 방이 없는데, 어디로 들어가요
```

```python
# 다음을 만족하는 함수를 작성하시오
# 1. space를 기준으로 word 를 구분
# 2. 3글자 이하는 제외
# 3. 가장 많이 나오는 단어 top 10 을 return 해주는 함수
# 예시
# top_ten_word(msg)
# ['will', 'that', 'have' ....]
```

```python
## 배운 것 위주로 사용 ##

def top_ten_words(msg):

# 1단계: 텍스트를 공백 기준으로 쪼개고 3글자 초과인 단어만 리스트에 담기

  words = msg.split()
  f_words = [word for word in words if len(word) > 3]

# 2단계: 단어별 빈도수 계산 (딕셔너리 활용)

  word_counts = {}
  for word in f_words:
    word_counts[word] = word_counts.get(word, 0) + 1    # 단어가 없으면 0, 있으면 기존 값에 1 추가
  #print(word_counts)                                  # 글자수 세기 완성

  top_10 = []                                         # top10 순위 리스트 만들기
  last_word = 0                                       # 공동순위 대비

# 3단계: 상위 10번 반복하며 가장 많이 나온 단어 추출

  for _ in range(10):                                 # for문을 10번 반복할 건데 이유는 top10 이니까. 이후 del 을 적용해서 가장 많은 것을 딕셔너리에서 삭제 하는 걸로 사용 -> 10번한다? 10번 돌리고, 많은 순으로 골라냈다.
    most_cnt = ""                                         #a 한번 돌렸을 때 가장 많은 단어
    now_cnt = 0                                           #b 작업반복문 돌릴 때 참고하라는 수치

    # for 문으로 가장 많이 찾아지는 단어를 찾기(딕셔너리에서 데이터를 참고)

    for mcserch in word_counts:
      if word_counts[mcserch] > now_cnt:               # 딕셔너리를 찾으면서 now_cnt 보다 클 경우
         now_cnt = word_counts[mcserch]                  #b 큰 것이 최고점으로 변경
         most_cnt = mcserch                                #a 그리고 그것이 가장 많은 요소로 선정

    top_10.append(most_cnt)                            # 가장 큰 요소를 top_10 에 넣어줌
    last_word = now_cnt                                  # 마지막 단어의 점수 기록
    del word_counts[most_cnt]                          # 넣은 이후에는 기존에 있던 딕셔너리에서 제외하여 기존 1위는 고려하지 않게끔 만들기

# 4단계: 마지막으로 10등과 점수가 같은 동점일 경우 구제

  for key in word_counts:
        if word_counts[key] == last_word:
            top_10.append(key)
            print(f" 공동순위 : {key}({last_word}회)")   # 공동 표시

  return top_10

result = top_ten_words(msg)

print(result)
```
