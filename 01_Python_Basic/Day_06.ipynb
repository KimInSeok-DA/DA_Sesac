# 굿모닝 코딩 : 크기가 작은 부분 문자열

숫자로 이루어진 문자열 t와 p가 주어질 때,

t에서 p와 길이가 같은 부분문자열 중에서,

이 부분문자열이 나타내는 수가 p가 나타내는 수보다 작거나 같은 것이 나오는 횟수를 return하는 함수 solution을 완성하세요.

예를 들어, t="3141592"이고 p="271" 인 경우,

t의 길이가 3인 부분 문자열은 314, 141, 415, 159, 592입니다.

이 문자열이 나타내는 수 중 271보다 작거나 같은 수는 141, 159 2개 입니다.

생각 순서
1. P의 문자열의 길이 -> len
2. T의 문자열을 len만큼 비교하되, [한칸씩 밀려나게끔] <- 이걸 어찌하오
3. 비교한 것이 P보다 작거나 같으면 answer에 +1
4. answer에 담겨진 개수 도출

```python
# 풀이 이전 range 복습
print(list(range(0,10,2)))    # 0~9까지인데, 2만큼 간격이 있어.
print(list(range(0,10)))      # 한칸씩 띄어야한다면 세번째는 공백으로 유지

t='3141592'                   # 7개   -> 314,131,159,592
t1='500220839878'             # 12개  -> 500,002, 022, .....
p="271"                       # 3개씩 ->
p1="7"                        # 1개씩 ->
n=len(p)                      # len(t)-n+1 : t의 길이 - p의 길이 +1 // +1 의 이유는 까지는 반영안하니까.

print(f"한칸씩 간다면 {list(range(0,len(t)-n+1))}")

def solution1(t, p):
  answer = []
  n=len(p)
  for i in range(0, len(t) - n + 1): # ai 사용                                                              # i 라는 작업을 할건데. i는 한칸씩 진행하는 것
    if int(t[i:i+n]) <= int(p) :     # 처음엔 일단 했지만 오류가 나서 오류값을 ai 물어봐서 수정함               # t의 [첫번째 자리부터 p의 길이만큼의 값] 한칸씩 진행 하는 값들이 p보다 작거나 같으면
      answer.append(t[i:i+n])                                                                               # 그 값을 answer에 넣음
  return answer
result = solution1(t, p)
print(result)
```

```python
# 나의 풀이

t='3141592'
t1='500220839878'
p="271"
p1="7"

def solution1(t, p):
  answer = 0                           # 문제에선 개수만 물어봄
  n=len(p)
  for i in range(0, len(t) - n+1):
    if int(t[i:i+n]) <= int(p) :
      answer +=1
  return answer
result = solution1(t, p)
print(result)
```

```python
# 실습 1 원을 클래스로 표현

import math
class Circle :
  def __init__ (self, radius=1.0):
    self.__radius = radius
  def setRadius(self, r):
    self.__radius = r
  def getRadius(self):
    return self.__radius
  def calcArea(self):
    area = math.pi*self.__radius**2
    return area
  def calcCircle(self):
    circumference = math.pi*self.__radius*2.0
    return circumference

c1 = Circle(10)
print("반지름=", c1.getRadius())
```

```python
# 실습 2 은행계좌를 클래스로 표현

class BankAccount :
  def __init__(self, balance=0):
    self.__balance = balance
  def withdraw(self, amount):
    self.__balance -= amount
    print(f"통장에서 {amount}원이 인출되었습니다.")
    return self.__balance
  def deposit(self, amount):
    self.__balance += amount
    print(f"통장에 {amount}원이 입금되었습니다.")
    return self.__balance

a = BankAccount()
a.withdraw(100)
a.deposit(10)
```

```python
# 실습 3 유클리디안 거리 # 교수님 풀이
pa = [2,3]
pb = [[2,5], [5,6], [6,7]]

def udistance(pa,pb):                                        # udistance 라는 함수를 만들건데, pa값(기준점)과 pb값(비교값)을 사용할거야
  dst = []
  sp = []
  for p in pb :                                              # pb 에서 끌어온 p값을 사용할 건데.
    dst.append(((p[0]-pa[0])**2 + (p[1]-pa[1])**2)**0.5)     # p값의 거리를 dst에 넣을거야.
  print(dst)                                                 # dst 는 pb의 요소별 거리 값
  sp = pb[dst.index(sorted(dst)[0])]                         # sp 는 정렬한 dst 값의 첫번째 자리의 위치값에 대해서 pb 값을 sp에 넣음   # sorted : 오름차순 정렬
  dst = sorted(dst)[0]                                       # dst를 정렬한 첫번째 값(가장 작은 값)
  return sp, dst
print(udistance(pa,pb))
```

만약 내 3차원 데이터의 형태가 (2,3,4) 라면 axis 가 지정한 shape의 위치를 제외하고 표시

x.sum(axis=0) -> 0자리에 있는 2가 지워져서 결과는 (3, 4) 형태의 2차원 표

x.sum(axis=1) -> 1자리에 있는 3이 지워져서 결과는 (2, 4) 형태의 2차원 표

x.sum(axis=2) -> 2자리에 있는 4가 지워져서 결과는 (2, 3) 형태의 2차원 표
