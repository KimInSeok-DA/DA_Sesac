```python
# 실습 6 glob 기능

import glob                           # 파일 읽어들이는 기능
glob.glob("*.txt")                    # txt로 된 확장자로 된 모든 파일 조회
```

```python
# 실습 6-1 glob 을 통한 출력

for fn in glob.glob("*.txt"):         # 모든 텍스트파일 하나씩 가져오기
  print(f"======={fn}=======")        # 뭘 가져왔는지 확인
  with open(fn) as f :                 # 그걸 열어서(완료 후 닫음)
    for line in f.readlines():           # 한줄씩 가져와서
      print(line)                        # 출력하기
```

```python
# 실습 6-2 glob을 통한 word counting 하기

wrdcnt = {}
with open("ihaveadream.txt") as f :
  for line in f.readlines():
    for word in line.split():
      wrdcnt[word] = wrdcnt.get(word,0) +1   # if, else 로도 가능
print(wrdcnt)
```

# text 파일을 읽어서 그 단어는 읽지 않게끔 제약조건 부여

```python
# 집합을 통한 풀이

wrdcnt = {}
with open("stopwords.txt") as st :
  stop_words = set(st.read().split())       # stopwords 에서 단어들 뽑아내서 담기
#print(stop_words)
with open("ihaveadream.txt") as f :
  for line in f.readlines():
    for word in line.split():
      if word in stop_words :                # 그 단어가 stop_words에 있어 넘어가기
        continue
      wrdcnt[word] = wrdcnt.get(word,0) +1   # if, else 로도 가능
print(wrdcnt)
```

```python
# 딕셔너리를 통한 풀이
stpwrd = {}
with open("stopwords.txt") as f:
  for line in f.readlines():
    for word in line.split():
      stpwrd[word.lower()] = 1
print(stpwrd)

wrdcnt = {}
with open("ihaveadream.txt") as f :
  for line in f.readlines():
    for word in line.split():
      if stpwrd.get(word.lower(), 0) == 0 :
        wrdcnt[word.lower()] = wrdcnt.get(word.lower(),0) +1
print(wrdcnt)
```

# Lambda

```python
# 실습 8 Lambda 사용

def f(x):                # 기존의 함수 사용 시
  return x+2
print(f(10))
```

```python
# 실습 8-1 Lambda 왜 사용?
lambda x: x+2                                                       # 일회성 함수를 만들때 사용
lst1 = sorted(wrdcnt.items(), key = lambda x:x[1], reverse= True)   # key = 정렬기준 설정 : 임시함수 "x를 입력하면 x의 두번째 위치"를 기준으로 정렬해
print(lst1)
```

```python
# 실습 8-2 Lambda를 활용한 다양한 정렬 방법(1)

lst1 = [["c",5,1],["a",3,1],["c",2,2],["d",1,4],["d",1,3]]
lst2 = sorted(lst1)
lst3 = sorted(lst1, key = lambda x:[x[2],x[0],x[1]])                # 위치에 의한 정렬 순서 조정 가능
print(lst2)
print(lst3)
```

```python
# 실습 8-2 Lambda를 활용한 다양한 정렬 방법(2)

lst1 = [[1,6,8],[1,5,9],[2,4,9],[2,6,9],[3,5,7],[3,5,9]]
print(sorted(lst1))
print(sorted(lst1, key = lambda x:(x[1],x[2],x[0])))
print(sorted(lst1, key = lambda x:(x[2],x[0],x[1])))
```

```python
# 실습 8-3 map 과 lambda

lst1 = ["abc", "abb", "bad", "bbb", "bca"]
lst2 = list(map(str.upper, lst1))              # lambda 없이 map 과 사용 시

print(lst2)
```

```python
# 실습 8-3 map 과 lambda 적용

lst1 = [1,2,3,4]
lst2 = list(map(lambda x:x**2+1, lst1))         # 숫자 계산
lst3 = list(map(lambda x:str(x)+'주', lst1 ))   # 숫자를 문자형으로 변환 후 문자 추가
lst4 = list(map(lambda x:int(x[0]), lst3 ))     # 일부 문자형을 숫자로 변환 저장
print(lst2)
print(lst3)
print(lst4)
```

```python
# 실습 9 word counting 적용                                                  # 기억하면 유용함

stpwrd = {}
with open("stopwords.txt") as f:
  for line in f.readlines():
    for word in line.split():
      stpwrd[word.lower()] = 1
print(stpwrd)

wrdcnt = {}
with open("ihaveadream.txt") as f :
  for line in f.readlines():
    for word in line.split():
      if word.isalpha():
        if stpwrd.get(word.lower(), 0) == 0 :
          wrdcnt[word.lower()] = wrdcnt.get(word.lower(),0) +1
      else:
        print(word, end="| ")                                              # end 기본값이 \n 임
print(wrdcnt)

topten = sorted(wrdcnt.items(), key = lambda x:x[1], reverse= True)[:10]
print("\n", topten)
```
