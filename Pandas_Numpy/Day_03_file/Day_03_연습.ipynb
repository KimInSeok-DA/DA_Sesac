### (10) 데이터에 정규화 또는 표준화 적용

```python
# 최소-최대 정규화
# 값의 범위를 0~1 사이로 맞춤
def minmax_normalize(arr):
    return (arr - arr.min()) / (arr.max() - arr.min())

# Z 점수 표준화
# 평균 0, 표준편차 1 기준으로 변환
def zscore_standize(arr):
    return (arr - arr.mean()) / arr.std()
```

```python
# TODO: 수치형 컬럼만 선택하고 그래프로 확인하기
cnum = concrete.iloc[:,:-2]
cnum.plot()
```

```python
# TODO: 최소-최대 정규화 적용하고 그래프로 확인하기
cnum.apply(minmax_normalize).plot()
```

```python
# TODO: Z 점수 표준화 적용하고 그래프로 확인하기
cnum.apply(zscore_standize).plot()
```

### 선택 미션

- `strength`를 `low / medium / high` 3단계로 나누기
- `label`별 `strength` 평균 계산하기
- `cement`와 `strength`의 관계를 산점도로 확인하기


### 1. `strength`를 `low / medium / high` 3단계로 나누기

```python
labels = ['low', 'medium', 'high']
concrete['strength_level'] = pd.cut(concrete['strength'], bins=3, labels=labels)
concrete.head()
```

```python
concrete['strength_level'].value_counts()
```

###2. `label`별 `strength` 평균 계산하기

```python
concrete.pivot_table(index='label',
                     values='strength',
                     aggfunc='mean',
                     observed=True)
```

### 3. `cement`와 `strength`의 관계를 산점도로 확인하기

```python
import matplotlib.pyplot as plt
concrete.plot(kind='scatter', x='cement', y='strength', color='purple', alpha=0.6, figsize=(8,6))
plt.title('Cement-Strength Scatter')

plt.show()
```
