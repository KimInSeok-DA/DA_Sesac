## 7) 수치 데이터를 범주형으로 변환

```python
concrete.head()
```

```python
labels=['weak', 'medium', 'strong', 'very strong']
concrete['label'] = pd.cut(concrete.strength, 4, labels=labels) # 4구간으로 나눠서 labels를 분류
concrete.head()
```

```python
pd.cut(concrete.strength, 4, labels=labels).value_counts() # 수치값을 기준으로만 4등분
```

## 수치 데이터를 표본 변위치 기준 범주형으로 변환

```python
pd.qcut(concrete.strength, 4, labels=labels) # 4개 비율로 나눠줌
```

```python
pd.qcut(concrete.strength, 4, labels=labels).value_counts()  # 비율 맞춰서 분류됨
```

## 10) 데이터에 정규화 또는 표준화 적용

```python
# 최소-최대 정규화
def minmax_normalize(arr):
    return (arr - arr.min())/(arr.max() - arr.min())  # 0~1 사이로 축소
```

```python
sample = np.arange(9).reshape(3,3)
sample
```

```python
minmax_normalize(sample)
```

```python
concrete.plot() # 원본 데이터
```

```python
cnum = concrete.iloc[:, :-2] # strngth 의미 없는 데이터라 지우고 적용
cnum.plot()
```

```python
cnum.apply(minmax_normalize).plot() # 최소-최대 정규 적용한 그래프
```

```python
# Z 점수 표준화              # 통계분석 3일차(#실습 5-1)
def zscore_standize(arr):
    return (arr - arr.mean())/arr.std()
```

```python
zscore_standize(sample)     # 표준화(평균 0, 표준편차 1)
```

```python
cnum.apply(zscore_standize).plot() # 표준화
```

## 11) 문자열 처리

```python
name = pd.DataFrame(['kim', 'Kim', 'lee', 'Lee'], columns=['lastname'])
name
```

```python
name.map(str.upper) # dataframe 을 통해 대문자 적용 / 2차원데이터
```

```python
name.lastname.map(str.upper) # series 을 통해 대문자 적용 / 1차원 데이터
```

```python
substitude_dict = {'Kim' : 'KIM',  # Series는 map 사용 을 통해 대문자 적용
                  'KIM' : 'KIM',
                  'kim' : 'KIM',
                  'Lee' : 'LEE',
                  'lee' : 'LEE'}

name.lastname.map(substitude_dict) # 딕셔너리를 적용시킬 수 있음
```
