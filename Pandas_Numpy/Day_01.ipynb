# 숫자 문자열과 영단어

네오와 프로도가 숫자놀이를 하고 있습니다. 네오가 프로도에게 숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"

234567 → "23four5six7"

10203 → "1zerotwozero3"

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나, 혹은 바뀌지 않고 그대로인 문자열 s가 매개변수로 주어집니다.

s가 의미하는 원래 숫자를 return 하도록 solution 함수를 완성해주세요.

참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

0	zero

1	one

2	two

3	three

4	four

5	five

6	six

7	seven

8	eight

9	nine

# 사고과정

1. 딕셔너리 구조같은데? + 리스트와 len으로도 가능하지 않을까?
2. 하나씩 바꿔야 하니 for 문으로
3. 그럼 변환하는 건 어떻게?? : replace 기능을 사용
4. 숫자열을 원하니 숫자열로 표기

```python
# 풀이 1
s = "one4seveneight"

def solution(s):
    num_dict = {"zero": "0", "one": "1", "two": "2", "three": "3", "four": "4", "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9"}

    for word, num in num_dict.items():   # 딕셔너리에서 요소를 하나씩 뽑아오기
        s = s.replace(word, num)         # replace("찾을단어", "바꿀단어")(+문자열만 됨)
    return int(s)                        # int 여야 문제가 풀림

print(solution(s))
```

```python
# 풀이 2
s = "one4seveneight"

def solution(s):
    nums = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

    for i in range(len(nums)):           # nums의 길이(10개) 만큼 작업 진행
        s = s.replace(nums[i], str(i))   # nums 리스트에 있는 것들을 i = len(nums) 숫자로 바꿈(문자열)
    return int(s)                        # 문자열이니 숫자로 변경

print(solution(s))
```

# Generator

한번에 한 결과를 만들어냄

""Return one Fibonacci element at a time ""

iterator 라는 객체 반환함

[] 대신 () 로 표현

```python
# 제너레이터

def fib_gen(n):
  a,b=0,1
  while b<n :
    yield b               # b를 출력하고 다음 진행하고 여기부터 다시 진행
    a,b = b, a+b

for x in fib_gen(1000) :
  print(x, end=', ')
```

for x in fib_gen(1000): 이 시작되면 다음과 같은 일이 벌어집니다.

0. 초기화: a = 0, b = 1이 됩니다. b < 1000 조건이 맞으므로 루프 진입!

1. 첫 번째 루프: yield b를 만납니다. 현재 b인 1을 for문의 x로 던져주고 함수는 그 자리에 일시 정지합니다.

   print(x)가 실행되어 화면에 1이 찍힙니다.

2. 두 번째 루프: for문이 다음 값을 달라고 요청하면, 멈췄던 yield b 바로 다음 줄부터 시작합니다.

    a, b = b, a + b 가 실행되어 a = 1, b = 1이 됩니다.

   다시 while 조건 검사 후 yield b를 만나서 1을 던져주고 또 일시 정지!

3. 세 번째 루프: 다시 깨어나서 a, b = b, a + b를 합니다. a = 1, b = 2가 됩니다.

    yield b를 만나서 2를 던져주고 일시 정지!

4. 이후 반복: 이 과정을 거쳐 3, 5, 8, 13 ... 순으로 값이 하나씩 밖으로 나갑니다.

5. 종료: b가 1000보다 커지는 순간(여기서는 987 다음 숫자인 1597이 될 때) while b < n 조건이 깨지면서 함수가 자연스럽게 종료되고 for문도 끝납니다.
