## Communities and Crime

**(1) 개요**


- 미국 내 여러 지역 공동체(Community)의 범죄율이 다양한 사회경제적 특성과 어떤 관련이 있는지를 분석하기 위해 만들어짐
- 데이터는 크게 세 가지의 공신력 있는 아래 데이터를 결합하여 구성됨
    - 1990년 미국 인구조사 (US Census): 인구, 인종, 소득, 교육 수준 등 사회경제적 데이터
    - 1990년 미국 법 집행 관리 및 행정 통계 (LEMAS): 경찰 인력, 예산 등 법 집행 관련 데이터
    - 1995년 미국 FBI 균일 범죄 보고서 (UCR): 살인, 강도, 폭행 등 실제 범죄 데이터
